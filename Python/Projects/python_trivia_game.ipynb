```python
# list of questions - done
# store the answers - done
# randomly pick questions
# ask the questions
# see if they are correct or not
# track the score
# tell the user their choice

import random

questions = {
    "What is the keyword to define a function in Python?": "def",
    "Which data type is used to store True or False values?": "boolean",
    "What is the correct file extension for Python files?": ".py",
    "Which symbol is used to comment in Python?": "#",
    "What function is used to get input from the user?": "input",
    "How do you start a for loop in Python?": "for",
    "What is the output of 2 ** 3 in Python?": "8",
    "What keyword is used to import a module in Python?": "import",
    "What does the len() function return?": "length",
    "What is the result of 10 // 3 in Python?": "3"
}

def python_trivia_game():
    questions_list = list(questions.keys()) #question
    total_questions = 5 #number of question for the user to answer
    score = 0 # initial score is 0

    selected_questions = random.sample(questions_list, total_questions) # to randomly select the questions any 5 questions

    for idx, question in enumerate(selected_questions):     #adding indexing function
        print(f'{idx+1}. {question}')
        user_answer = input("Your answer: ").lower().strip()   # adding user input to answer the questions

        correct_answer = questions[question]   # defining the correct answer

        if user_answer == correct_answer.lower():  # giving lower case to the user input answer to overcome the case sensitivity for the user answer input
            print("Correct!\n")   # if the answer is correct, print as correct and go to the next line
            score += 1            # if the answer is correct, add one mark to the score
        else:
            print(f'Wrong. The correct answer is: {correct_answer}.\n')  # if the answer is wrong, print as wrong and then  the correct answer

    print(f'Game over! your final score is: {score}/{total_questions}')  # post completing the loop print as game over and display the final score

python_trivia_game()
```
