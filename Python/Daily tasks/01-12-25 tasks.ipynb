```python
# Task 1 - Create an abstract class Shape with an abstract method area().
# Implement Rectangle, Circle, and Triangle classes and calculate their areas.

from abc import ABC, abstractmethod
class shape(ABC):
    @abstractmethod
    def area(self):
        pass

class rectangle(shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class circle(shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius**2)

class triangle(shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

rect = rectangle(10,5)
cir = circle(8)
tri = triangle(5,15)

print(f'Area of rectange: {rect.area()}cm')
print(f'Area of circle: {cir.area()}cm')
print(f'Area of triangle: {tri.area()}cm')
```

```python
# Task 2 - Create three classes:
# User
# Admin(User)
# SuperAdmin(Admin)
# Each class should have at least one unique method. Demonstrate method overriding.

class user:
    def __init__(self,name):
        self.name = name

    def show_role(self):
        return f'I am User'

    def show_details(self):
        return f'User: {self.name}'

class admin(user):
    def __init__(self, name, admin_id):
        super().__init__(name)
        self.admin_id = admin_id

    def show_role(self):
        return f'I am admin'

    def show_details(self):
        return f'Admin: {self.name}, ID: {self.admin_id}'

class superadmin(admin):
    def __init__(self, name, admin_id, level):
        super().__init__(name, admin_id)
        self.level = level

    def show_role(self):
        return f'I am SuperAdmin'

    def show_details(self):
        return f'SuperAdmin: {self.name}, ID: {self.admin_id}, Level: {self.level}'

user1 = user('Teja')
admin1 = admin('Sam', 100)
superadmin1 = superadmin('Ram', 101, 'Mid-level')

print(user1.show_role())
print(admin1.show_role())
print(superadmin1.show_role())
print()
print(user1.show_details())
print(admin1.show_details())
print(superadmin1.show_details())
```

```python
# Task 3 - Create a real-life Transport System:
# Abstract class: Transport â†’ method calculate_fare()
# Child classes: Bus, Auto, Train
# Use encapsulation for private distance and rate

from abc import ABC, abstractmethod
class transport(ABC):
    def __init__(self, rate):
        self.__rate = rate

    @abstractmethod
    def calculate_fare(self, distance):
        pass

    def get_rate(self):
        return self.__rate
        
class bus(transport):
    def calculate_fare(self, distance):
        self.__distance = distance
        return f'Bus fare: {self.__distance * self.get_rate()} INR'

    def get_distance(self):
        return self.__distance

class auto(transport):
    def calculate_fare(self, distance):
        self.__distance = distance
        return f'Auto fare: {self.__distance * self.get_rate()} INR'

    def get_distance(self):
        return self.__distance

class train(transport):
    def calculate_fare(self, distance):
        self.__distance = distance
        return f'Train fare: {self.__distance * self.get_rate()} INR'

    def get_distance(self):
        return self.__distance

bus1 = bus(5)
print(bus1.calculate_fare(31))
auto1 = auto(10)
print(auto1.calculate_fare(5))
train1 = train(0.2)
print(train1.calculate_fare(650))
```
