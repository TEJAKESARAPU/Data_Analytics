```python
# 1. Write a Python function to reverse a string without using slicing or reversed(). 

def reverse_string(strings):
    rev = ''
    for ch in strings:
        rev = ch + rev
    return rev
print(reverse_string('python'))
```

```python
# 2. Write a program to count the frequency of each character in a given string. 

def freq_char(characters):
    freq = {}
    for ch in characters:
        if ch in freq:
            freq[ch] +=1
        else:
            freq[ch] = 1
    return freq
print(freq_char('hello'))
```

```python
# 3. Remove duplicates from a list without using set(), while maintaining the original order. 

def remove_dup(string):
    result = []                 #we want the result in a list, so we used []
    for ch in string:
        if ch not in result:
            result.append(ch)
    return result
print(remove_dup('abbccddeff'))
```

```python
# 4. Write a Python function to find the second largest number in a list without using sort(). 

def second_largest(nums):
    largest = max(nums)
    nums.remove(largest)
    return max(nums)
print(second_largest([1,2,3,4,5,6,7,8]))
```

```python
# 5. Write a program to check whether two strings are anagrams of each other. 

def anagrams(str1, str2):
    if sorted(str1) == sorted(str2):
        return f'It\'s an Anagram!'
    elif len(sorted(str1)) != len(sorted(str2)):
        return f'Not an Anagram'
    else:
        return f'Not an Anagram'
print(anagrams('earth', 'heart'))
print(anagrams('listen', 'silent'))
print(anagrams('cat', 'act'))
```

```python
# 6. Write a recursive function to generate the first n Fibonacci numbers. 

n = int(input('Enter a number:'))
def fibonacci(n):
    if n <= 0:
        return []      #case 1 - if n = 0 or less than 0
    elif n == 1:
        return [0]     #case 2- if n = 1
    elif n == 2:
        return [0, 1]  #case 3 - if n =2

    seq = fibonacci(n-1)             #Genarate fibonacci series upto n-1
    seq.append(seq[-1] + seq[-2])    #Add the next number(sum of last two numbers)
    return seq
print(fibonacci(n))
```

```python
# 7. Using lambda, map, and filter, create a list of squares of even numbers. 

numbers = [1,2,3,4,5,6,7,8]
result = list(map(lambda n: n**2, filter(lambda n: n%2==0, numbers)))
print(result)
```

```python
# 10. Create a Bank Account class that supports: 
# deposit() 
# withdraw() 
# check balance 

class BankAccount:
    def __init__(self,initial_balance=0):
        self.balance = initial_balance

    def deposit(self, amount):
        if amount >= 100:
            self.balance += amount
            print(f'Deposited: {amount}, Current Balance: {self.balance}')
        else:
            print('Deposited amount must be >= 100')

    def withdraw(self, amount):
        if amount >= 100:
            if amount <= self.balance:
                self.balance -= amount
                print(f'Withdraw amount: {amount}, Current Balance: {self.balance}')
            else:
                print('Insufficient funds!!')
        else:
            print('Withdraw amount must be greather than or equal to 100')

    def check_balance(self):
        print(f'Current balance: {self.balance}')
        return self.balance

acc = BankAccount(10000)
acc.deposit(2000)
acc.deposit(1552)
acc.withdraw(500)
acc.check_balance()
```

```python
# 11. Write an example of method overriding using inheritance (polymorphism). 

class animal:
    def sound(self):
        pass

class dog(animal):
    def sound(self):
        print('Bark')

class cat(animal):
    def sound(self):
        print('Meow')

class cow(animal):
    def sound(self):
        print('Moo')

animals = [dog(), cat(), cow()]
for animal in animals:
    animal.sound()
```
