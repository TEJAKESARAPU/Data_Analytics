```python
# #class task
# create your record - name, city, course, college (write, read, append, etc, (all the operations))

with open('record.txt', 'w') as f:      #creating a file
    f.write('Name: Teja\n')             #create
    f.write('City: Visakhapatnam\n')    #create
```

```python
with open('record.txt', 'a') as f:
    f.write('Course: Mechanical Engineering\n')     #adding
    f.write('College: Raghu Engineering College\n')   #adding
```

```python
with open('record.txt', 'r') as f:
    print(f.read())                   #reading the file
```

```python
with open('record.txt', 'r') as f:
    print(f.read(10))                  #reading the file by character
```

```python
with open('record.txt', 'r') as f:
    print(f.readline().strip())                   #reading the file line by line
    print(f.readline().strip())                   #reading the file line by line
```

```python
with open('record.txt', 'r') as f:
    print(f.readlines())                   #reading all the lines as a list
```

```python
with open('record.txt', 'r') as f:
    for line in f:
        print('>>', line.strip())       #looping
```

```python
with open('record.txt', 'r+') as f:
    data = f.read()
    f.write('This is my record' + '\n' + data)
    print(data)
```

```python
with open('record.txt', 'r') as f:
    print(f.read())
```

```python
with open('record.txt', 'r') as f:
    data = f.read()

with open('record_copy.txt', 'w') as f1:
    f1.write(data)
```

```python
with open('record_copy.txt', 'r') as f1:
    print(f1.read())
```
