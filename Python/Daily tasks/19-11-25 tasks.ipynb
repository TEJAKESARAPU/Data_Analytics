```python
# Task 1 - From a list of names, create a list of vowels found inside each name. 
# names = ["Mohit", "Aman", "Neha"]

names = ["Mohit", "Aman", "Neha"]
vowels = [[i for i in name if i in 'aeiouAEIOU'] for name in names]
print(vowels)
```

```python
# Task 2 - Given a sentence, create a dictionary where:
# sentence = "Python comprehension makes coding easier"
# key  =  word
# value  =  length of the word

sentence = "Python comprehension makes coding easier"
dict1 = {word : len(word) for word in sentence.split()}
print(dict1)
```

```python
# Task 3 - Given a list of words, extract the unique first letters using set comprehension.
# words = ["apple", "banana", "apricot", "cherry", "blueberry"] 

words = ["apple", "banana", "apricot", "cherry", "blueberry"] 
unique = {ch[0] for ch in words}
unique
```
