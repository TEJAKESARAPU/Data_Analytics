```python
# Task 1 - Create a class Product with price & discount.
# Add a method to calculate the final price after discount.

class product:
    def __init__ (self, name, original_price, discount_percentage):
        self.name = name
        self.original_price = original_price
        self.discount_percentage = discount_percentage

    def final_price(self):
        final_price = self.original_price * (1-self.discount_percentage/100)
        return f'{self.name} price after discount is Rs.{final_price}'

pr1 = product("Book", 200, 10)
print(pr1.final_price())
pr2 = product('T-shirt', 2200, 30)
print(pr2.final_price())
```

```python
# Task 2 - Create a Student class to store name and marks.
# Write a method to return PASS or FAIL (pass if marks â‰¥ 40)

class student:
    def __init__ (self, name, marks):
        self.name = name
        self.marks = marks

    def results(self):
        if self.marks >= 40:
            return f'{self.name} has pass the test'
        else:
            return f'{self.name} has Failed the test'

stu1 = student('Teja', 95)
stu2 = student('Ram', 38)
print(stu1.results())
print(stu2.results())
```

```python
# Task 3 - Create a class BankAccount with deposit and withdraw methods (It should print insufficient amount or withdraw amount).
# Balance should never go negative.

class BankAccount:
    def __init__ (self, acc_holder_name, balance):
        self.acc_holder_name = acc_holder_name
        self.balance = balance

    def deposit(self, deposit_amount):
        self.deposit_amount = deposit_amount
        if deposit_amount > 0:
            self.balance += deposit_amount
            return f'Deposit amount: Rs.{self.deposit_amount}, Total balance: Rs.{self.balance}'
        else:
            return f'Negative amount'

    def withdrawal(self, withdrawal_amount):
        self.withdrawal_amount = withdrawal_amount
        if self.withdrawal_amount < self.balance:
            self.balance -= withdrawal_amount
            return f'Withdrawal amount: Rs.{self.withdrawal_amount}, Total balance: Rs.{self.balance}'
        else:
            return f'insufficient balance'

acc1 = BankAccount('Teja', 15000)
print(acc1.deposit(500))

acc2 = BankAccount('Sam', 2250)
print(acc2.withdrawal(150))
```
