```python
''' Task 2 - 

Input marks from the user and print:
"Pass" if marks is greater than equal to 40
"Fail" otherwise '''

marks = int(input('Enter your marks: '))
if marks >= 40:
    print('Pass')
else:
    print('Fail')
```

```python
''' Task 3 - 

If a person drinks 2 litres of water per day, write a program to calculate the total water consumed in a week.
Then compare if itâ€™s greater than or equal to 14 litres '''

water_consumed_per_day = int(input('How many litres water did you consume today?:  '))
water_consumed_in_a_week = water_consumed_per_day * 7

if water_consumed_per_day >= 2 and water_consumed_in_a_week >= 14:
    print('You have consumed:', water_consumed_in_a_week, 'litres of water this week')
    print('You are healthy!!!')
else:
    print('You have consumed:', water_consumed_in_a_week, 'litres of water this week')
    print('Please drink more water')
```
