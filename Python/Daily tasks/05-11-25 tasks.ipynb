```python
# Task 2 - Given the list:
# colors = ['red', 'green', 'blue', 'yellow', 'pink', 'black']
# Write Python statements to:
# Print the first three colors
# Print all colors except the first two
# Print every alternate color

colors = ['red', 'green', 'blue', 'yellow', 'pink', 'black']
print('First three colors are', ', '.join(colors[0:3:1]))
print('All colors except the first two colors are', ', '.join(colors[2::]))
print('Alternative colors are', ', ' .join(colors[::2]))
```

```python
# Task 3 - Create a list of your favorite foods:
# foods = ['pizza', 'burger', 'pasta', 'momos']
# Perform the following operations and print results:
# Add two more items using .append()
# Insert an item at position 2
# Remove one item using .remove()
# Reverse the list
# Sort the list alphabetically

foods = ['Pizza', 'Burger', 'Pasta', 'Momos']
foods.append('Dosa')
foods.append('Vada')
print('After performing append operation:', foods)
foods.insert(2, 'Ice cream')
print('After insrting Ice cream in postion 2:', foods)
foods.remove('Pizza')
print('After removing Pizza from the foods:', foods)
foods.reverse()
print('Reversing the foods list:', foods)
foods.sort()
print('Sorting the foods list:', foods)
```

```python
# Task 4 - Create a list cart = ['milk', 'bread', 'eggs']
# Perform the following operations:
# Add 'butter' and 'jam' to your cart.
# Remove 'bread'.
# Print the total number of items in your cart.
# Display all items in the cart, one per line.

cart = ['milk', 'bread', 'eggs']
cart.append('butter')
cart.append('jam')
print('Added butter and jam to the cart', cart)
print()
cart.remove('bread')
print('Removed bread from the cart:', cart)
print()
print('Total number of items in the cart are', len(cart), 'items')
print()
print('Items in cart are as follows: \n')
for item in cart:
    print(item)
```
