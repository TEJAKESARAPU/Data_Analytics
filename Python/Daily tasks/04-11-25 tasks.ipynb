```python
''' Task 1 - Write a Python program to calculate the area and perimeter of a rectangle '''

length = float(input('Enter length of the rectangle:'))
width = float(input('Enter width of the rectangle:'))

Area = length * width
perimeter = 2 * (length + width)

print('Area of the rectangle is ', Area)
print('perimeter of the rectangle is ', perimeter)
```

```python
''' Task 2 - Compute and print the simple interest:
Formula → SI = (P * R * T) / 100  (Ask user to input) '''

#values
p = float(input('Principal amount: '))
r = float(input('Rate of Interest: '))
t = float(input("Time(in years) "))

#Calculating Sipmple interest
si = (p*r*t) / 100

print("Simple interest: ", si)
```

```python
''' Task 3 - Suppose you’re tracking daily steps.
Initialize steps = 0
Add the following daily steps: 3200, 4500, 6000 '''

steps = 0
steps += 3200
steps += 4500
steps += 6000

print('Milestone achieved: ', steps, 'Steps')
```

```python
''' Task 4 - Write a program that takes two numbers and prints "Both even" if both are even, otherwise "Not Second even". '''

num1 = int(input('First number: '))
num2 = int(input('Second number: '))

if num1 % 2 == 0 and num2 % 2 == 0:
    print('Both are even numbers')
else:
    print('Not second even')
```
