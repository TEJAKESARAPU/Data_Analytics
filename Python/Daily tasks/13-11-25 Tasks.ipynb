```python
#task 1
#take input as user input and count vowels

name = input('Enter your name:').lower() #TAKEN USER INPUT AND MADE THE INPUT AS LOWER CASE
count = 0  
vowels = 'aeiou'
for char in name:
    if char in vowels:
        count +=1
print('No. of vowels for the name', name, 'is', count)
```

```python
#Task 2
# Create an inventory and take an order as user input, 
# then check whether the order is available in the inventory or not

inventory = ['apple', 'grapes', 'hot chocolate', 'veg puff', 'samosa', 
             'cookies', 'lays', 'butter', 'milk', 'banana', 
             'mirror', 'soap', 'brush', 'paste', 'white pasta', 'sushi']

# Take order input (multiple items separated by commas)
order = input('Enter your order items (comma separated): ').lower().split(',')

# Strip spaces around each item
order = [item.strip() for item in order]

# Check availability
for item in order:
    if item in inventory:
        print(f"✅ {item} is available")
    else:
        print(f"❌ {item} is not available")
```
