```python
# Q.1- Create a circle class and initialize it with a radius. Make two methods, getArea and getCircumference, inside this class.

import math
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def getArea(self):
        return f'Circle radius: {math.pi * self.radius * self.radius}'

    def getCircumference(self):
        return f'Circle circumference: {2 * math.pi * self.radius}'

cir = Circle(5)
print(cir.getArea())
print(cir.getCircumference())
```

```python
# Q.2- Create a Student class and initialize it with name and roll number. Make methods to:
# Display - It should display all informations of the student.
# setAge - It should assign age to student
# setMarks - It should assign marks to the student.


class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number

    def display(self):
        return f'Name: {self.name}\nRoll number: {self.roll_number}\nAge: {self.age}\nMarks: {self.marks}'

    def setAge(self, age):
        self.age = age

    def setMarks(self, marks):
        self.marks = marks

s1 = Student('Teja', 74)
s1.setAge(24)
s1.setMarks(94)
print(s1.display())
```

```python
# Q.3- Create a Temperature class. Make two methods:
# convertFahrenheit - It will take Celsius and print it in Fahrenheit.
# convertCelsius - It will take Fahrenheit and convert it into Celsius.

class Temperature:
    def convertFahrenheit(self, celsius):
        self.celsius = celsius
        fahrenheit = (self.celsius * (9/5)) + 32
        return f'Celsius: {self.celsius}°, Fahrenheit: {fahrenheit}°'           #F = (°C × 9/5) + 32

    def convertCelsius(self, fahrenheit):
        self.fahrenheit = fahrenheit
        celsius = (self.fahrenheit - 32) * (5/9)
        return f'Fahrenheit: {self.fahrenheit}°, Celsius: {celsius}°'

cel = Temperature()
print(cel.convertFahrenheit(35))
print(cel.convertCelsius(77))
```

```python
# Q.5- Create a class Expenditure and initialize it with salary, savings, category, and total expenditure. Make the following methods.
# Add expenditure according to category.
# Calculate total expenditure.
# Calculate per day and monthly expenditure.

class Expenditure:
    def __init__(self, salary, savings):
        self. salary = salary
        self.savings = savings
        self.category_expenses = {}
        self.total_expenditure = 0

    def add_expenditure(self, category, amount):
        if category in self.category_expenses:
            self.category_expenses[category] += amount
        else:
            self.category_expenses[category] = amount

        self.total_expenditure += amount

    def calculate_total_expenditure(self):
        return self.total_expenditure

    def calculate_daily_monthly(self):
        daily = self.total_expenditure/30
        monthly = self.total_expenditure
        return daily, monthly

    def display(self):
        print(f'Salary: {self.salary}\n'
              f'Savings: {self.savings}\n'
              f'Category-wise Expenditure: {self.category_expenses}\n'
              f'Total Expenditure: {self.total_expenditure}')

exp = Expenditure(67000, 15000)

exp.add_expenditure('Food', 4000)
exp.add_expenditure('Fuel', 1600)
exp.add_expenditure('Shopping', 2800)
exp.display()

daily, monthly = exp.calculate_daily_monthly()
print(f'Per Day Expenditure: {daily}')
print(f'Per Month Expenditure: {monthly}')
```

```python
# Q.7- Create a class Cop. Initialise its name, age and work experience. Define methods to display and update the details. 
# Create another class Mission, which extends the class Cop.

class Cop:
    def __init__(self, name, age, experience):
        self.name = name
        self.age = age
        self.experience = experience

    def display_details(self):
            return f'Name: {self.name}, Age: {self.age}, Experience: {self.experience}'

    def update_details(self, name=None, age=None, experience=None):
            if name:
                self.name = name
            if age:
                self.age = age
            if experience:
                self.experience = experience

class Mission(Cop):
    def mission_details(self):
        return f'Mission assigned to cop: {self.name}'

cop1 = Cop('Ravi', 28, 2)
print(cop1.display_details())

cop1.update_details(age=29, experience=3)
print(cop1.display_details())

print()

m1 = Mission('Teja', 30, 5)
print(m1.display_details())
print(m1.mission_details())
```

```python
# Q.8- Define the method add_mission_details. Select an object of Cop and access the methods of the base class to get information for a particular cop 
# and make it available for the mission.

class Mission(Cop):
    def mission_details(self):
        pass

    def add_mission_details(self, cop_object, mission_name):
        cop_info = cop_object.display_details()
        return f'{mission_name} assigned.\nCop Details:\n{cop_info}'

cop2 = Cop('Arjun', 30, 5)
mission = Mission('Officer', 40, 15)
print(mission.add_mission_details(cop2, 'Recuse mission'))
```

```python
# Q.9- Create a class Shape. Initialise it with length and breadth. Create the method Area. 
# Create a class rectangle and square that inherits shape and access the method Area.

class Shape:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
        
    def area(self):
        pass

class Rectangle(Shape):
    def area(self):
        return f'Rectangle area: {self.length * self.breadth}'

class Square(Shape):
    def area(self):
        return f'Square area: {self.length * self.length}'

rect = Rectangle(4,6)
squ = Square(4,4)
print(rect.area())
print(squ.area())
```
