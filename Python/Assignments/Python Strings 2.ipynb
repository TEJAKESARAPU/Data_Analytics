```python
# Q1. Remove spaces before and after the string, convert to uppercase, and print the length of the final string. Also print the uppercase string.
#  Input: " Python Programming "
#  Output:
#  "PYTHON PROGRAMMING"
#  18

text = " Python Programming "
text = text.strip()
print(text.upper())
print(len(text))
```

```python
# Q2. Replace "a" with "A", then check if the final string ends with "A". Also print the modified string.
#  Input: "banana"
#  Output:
#  "bAnAnA"
#  True

text = 'banana'
text = text.replace('a', 'A')
print(text)
print(text[-1] == 'A')
```

```python
# Q3. Remove "www." and ".com", then print the remaining string. Also print the original and final versions.
#  Input: "www.username.com"
#  Output:
#  "www.username.com"
#  "username"

text = "www.username.com"
print(text)
tex = text.replace('www.', '').replace('.com', '')
print(tex)
```

```python
# Q4. Convert to title case and then check if the final string starts with "Python". Also print the title case string.
# Input: "Python is amazing"
# Output:
# "Python Is Amazing"
#  True

text = "Python is amazing"
text = text.title()
print(text)
print(text.startswith('Python'))
```

```python
# Q5. Remove "@", then check if the final string is alphanumeric. Also print the modified string.
#  Input: "Data@123"
#  Output:
#  "Data123"
#  True

text = "Data@123"
text = text.replace('@', '')
print(text)
print(text.isalnum())
```

```python
# Q6. Replace "_" with a space, then capitalize only the first word and print the result. Also, print the intermediate string after the replacement.
#  Input: "hello_world"
#  Output:
#  "hello world"
#  "Hello world"

text = "hello_world"
text = text.replace('_', ' ')
print(text)
print(text.capitalize())
```

```python
# Q7. Extract and print only the domain name from the email. Also, print the split parts before extraction.
#  Input: "user@gmail.com"
#  Output:
#  ['user', 'gmail.com']
#  "gmail"

text = "user@gmail.com"
parts = text.split('@')
print(parts)
domain = parts[1].split('.')[0]
print(domain)
```

```python
# Q9. Convert to lowercase, then capitalize and print the final string. Also print the lowercase version.
#  Input: "HELLO world"
#  Output:
#  "hello world"
#  "Hello world"

text = "HELLO world"
text = text.lower()
print(text)
text = text.capitalize()
print(text)
```

```python
# Q10. Convert only the file extension to lowercase and print the result. Also print the filename without modification.
#  Input: "File.PDF"
#  Output:
#  "File.PDF"
#  "File.pdf"

text = "File.PDF"
print(text)
parts = text.split('.')
part1 = parts[1].lower()
new_text = parts[0] + '.' + part1
print(new_text)
```

```python
# Q10. Convert only the file extension to lowercase and print the result. Also print the filename without modification.
#  Input: "File.PDF"
#  Output:
#  "File.PDF"
#  "File.pdf"

text = "File.PDF"
print(text)
print(text.split('.')[0] + "." + text.split('.')[1].lower())
```

```python
# Q11. Replace "5" with "FIVE", check if the result is alphanumeric, and print the answer. Also print the replaced string.
#  Input: "12345"
#  Output:
#  "1234FIVE"
#  True

num = "12345"
num = num.replace('5', 'FIVE')
print(num)
print(num.isalnum())
```

```python
# Q12. Take a string and count how many times the letter "o" appears.
# Input: hello world
# Output: 2

text = input('Enter a text:')
count = 0

for ch in text:
    if ch == 'o':
        count+=1
print(count)
```

```python
# Q14. Check if All Characters are Alphabetic or Digits
# Take a string and check: Is it alphabetic?, and Is it numeric?

text = input('Enter something:')
for ch in text:
    if ch.isdigit():
        print(f'{ch} is a number')
    elif ch.isalpha():
        print(f'{ch} is a alphabet')
    else:
        print(f'{ch} is neither alphabet nor number')

print(text ,"is alphabetic?", text.isalpha())
print(text, "is numeric?", text.isdigit())
```
