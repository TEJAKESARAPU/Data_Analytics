```python
# 4. Write a function that accepts a number and returns whether it is prime or not.
# Input: 7
# Output: True

def prime_num(num):
    if num <=1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(prime_num(7))
```

```python
# 5. Write a function that accepts a string and returns the number of vowels in it.
# Input: "programming"
# Output: 3

def count_vowels(string):
    vowels = 'aeiou'
    return sum(1 for ch in string.lower() if ch in vowels)
print(count_vowels('programming'))
```

```python
# 6. Write a function that accepts two numbers and returns their greatest common divisor (GCD).
# Input: a = 12, b = 18
# Output: 6

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
print(gcd(12,18))
```

```python
# 7. Write a function that accepts a list of numbers and returns the list after removing duplicates.
# Input: [1, 2, 2, 3, 4, 4, 5]
# Output: [1, 2, 3, 4, 5]

def remove_dup(nums):
    return list(set(nums))

print(remove_dup([1, 2, 2, 3, 4, 4, 5]))
```

```python
# 8. Write a function that accepts a dictionary and returns the key with the maximum value.
# Input: {'a': 10, 'b': 20, 'c': 5}
# Output: 'b'

def max_key(d):
    return max(d, key=d.get)
nums = {'a': 10, 'b': 20, 'c': 5}
print(max_key(nums))
```

```python
# 9. Write a function that accepts a number n and returns the factorial of n.
# Input: 5
# Output: 120

def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *=i
    return result
print(factorial(5))
```

```python
# 10. Write a function that accepts a string and returns a dictionary with the frequency of each character.
# Input: "apple"
# Output: {'a: 1, 'p': 2, 'l': 1, 'e': 1}

def char_frq(string):
    return {ch: string.count(ch) for ch in set(string)}
print(char_frq('apple'))
```

```python
# 15. Write a function that accepts three positional arguments and two keyword arguments, and returns a formatted string.
# Input: "Alice", 25, "Engineer", city="Hyderabad", country="India"
# Output: "Name: Alice, Age: 25, Profession: Engineer, City: Hyderabad, Country: India"

def stu_info(name, age, profession, **details):
    city = details.get('city')
    country = details.get('country')
    return f'Name: {name}, Age: {age}, Profession: {profession}, City: {city}, Country: {country}'

print(stu_info("Alice", 25, "Engineer", city="Hyderabad", country="India"))
```

```python
# 18. Write a function multiply(*numbers) that returns the product of all numbers passed to it. Call it with 3 and 5 arguments.
# Input: multiply(2, 3, 4)
# Output: 24
# Input: multiply(1, 2, 3, 4, 5)
# Output: 120

def multiply(*numbers):
    product = 1
    for num in numbers:
        product *= num
    return product

print(multiply(2, 3, 4))
print(multiply(1, 2, 3, 4, 5))
```

```python
# 19. Write a function concat_strings(*args) that joins all string arguments with a space and returns the result. Call it with 2 and 4 strings.
# Input: concat_strings("Hello", "World")
# Output: "Hello World"
# Input: concat_strings("Python", "is", "fun", "today")
# Output: "Python is fun today"

def concat_strings(*args):
    return ' '.join(args)
print(concat_strings("Hello", "World"))
print(concat_strings("Python", "is", "fun", "today"))
```

```python
# 23. Write a function print_kwargs(**kwargs) that prints all key-value pairs passed as keyword arguments. Call it with 1 and 3 key-value pairs.
# Input: print_kwargs(name="Alice")
# Output: name: Alice
# Input: print_kwargs(name="Bob", age=25, city="Delhi")
# Output:
# # name: Bob
# age: 25
# city: Delhi

def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f'{key} : {value}')

print_kwargs(name="Alice")
print()
print_kwargs(name="Bob", age=25, city="Delhi")
```

```python
# 24. Write a function greet_friends(*friends, msg="Hi") that prints the greeting message followed by each friend's name. Call it with 2 and 4 friends, with and without msg.
# Input: greet_friends("Alice", "Bob")
# Output:
# Hi Alice
# Hi Bob
# Input: greet_friends("Tom", "Jerry", "Spike", "Tyke", msg="Hello")
# Output:
# Hello Tom
# Hello Jerry
# Hello Spike
# Hello Tyke

def greet_friends(*friends, msg='Hi'):
    for friend in friends:
        print(f'{msg} {friend}')
greet_friends("Alice", "Bob")
print()
greet_friends("Tom", "Jerry", "Spike", "Tyke", msg="Hello")
```

```python
# 25. Write a function sum_numbers(*numbers, start=0) that returns the sum of all numbers plus start. Call it with numbers only and with start value.
# Input: sum_numbers(1, 2, 3)
# Output: 6
# Input: sum_numbers(1, 2, 3, start=5)
# Output: 11

def sum_numbers(*numbers, start=0):
    for i in numbers:
        return sum(numbers) + start
print(sum_numbers(1, 2, 3))
print(sum_numbers(1, 2, 3, start=5))
```

```python
# 27. Write a function list_to_string(*items, sep='-') that joins all items using the separator. Call it with default and custom separator.
# Input: list_to_string("a", "b", "c")
# Output: "a-b-c"
# Input: list_to_string("1", "2", "3", sep=":")
# Output: "1:2:3"

def list_to_string(*items, sep='-'):
    return sep.join(items)
print(list_to_string("a", "b", "c"))
print(list_to_string("1", "2", "3", sep=":"))
```
