```python
# # 3. Output Prediction Raising an Exception
# # What will be the output of the following program?

try:
    raise NameError('Hi there')
except NameError:
    print('An exception')
    raise

# Ans. First prints and then raises the user-defined error: NameError
```

```python
# 4. Output Prediction — Try–Except–Else Block
# What will be the output of the following program?

def AbyB(a,b):
    try:
        c=((a+b)/(a-b))
    except ZeroDivisionError:
        print('a/b results in 0')
    else:
        print(c)

AbyB(2.0, 3.0)
AbyB(3.0, 3.0)
```

```python
# 5. Program Writing — Handling Built-in Exceptions
# Write a Python program to demonstrate the handling of the following exceptions:
# ImportError
# ValueError
# IndexError

#ImportError
try:
    import non_existent_module
except ImportError as e:
    print('ImportError:', e)

#ValueError
try:
    number = int('abc')
except ValueError as e:
    print('ValueError:', e)

#IndexError
try:
    lst = [1,2,3,4,5]
    print(lst[7])
except IndexError as e:
    print('IndexError:', e)

```

```python
# 6. User-Defined Exception
# Create a user-defined exception called AgeTooSmallError that displays a warning message when the user enters an age less than 18.
# The program should repeatedly ask for input until the user enters a valid age (18 or above).


class AgeTooSmallError(Exception):
    pass

while True:
    try:
        age = int(input('Enter your age:'))
        
        if age < 18:
            raise AgeTooSmallError('Age is below 18. Please enter 18 or above')
        else:
            print('Valid age', age)
            break
    except AgeTooSmallError as e:
        print('AgeTooSmallError caught:',e)
    except ValueError:
        print('Invalid input! Please enter a number')
```
