```python
from teja_bank import BankAccount, InsufficientBalanceError

# Task 1: Create a new account (Account number: 1001, Account holder: YourName, Initial balance: 5000). Print the account details.

ba = BankAccount(0,'',0.0)
print(ba.create_account(1001, 'Kesarapu', 5000))
```

```python
# Task 2: Try to credit 2000 into the account.  
# Use try-except-else-finally.
# If successful, print "Credit successful".
# If error, print the error message.
# Always display "Transaction process completed".

try:
    ba.credit(2000)
except ValueError as e:
    print("Error:", e)
else:
    print("Credited successful")
finally:
    print("Transaction process completed")
```

```python
# Task 3: Try to debit 1000 from the account.
# Handle exceptions properly.
# If successful, print "Debit successful".
# Else, print the error message.

try:
    ba.debit(1000)
except (ValueError, InsufficientBalanceError) as e:
    print('Error:', e)
else:
    print('Debit successful')
```

```python
# Task 4: Try to debit 8000 from the account (more than the available balance).
# This should raise your InsufficientBalanceError.
# Handle it using try-except.

try:
    ba.debit(8000)
except InsufficientBalanceError as e:
    print('Error:', e)
```

```python
# Task 5: Try to credit a negative amount (-500).
# This should raise a ValueError.
# Handle it and print the error message.


try:
    ba.credit(-500)
except ValueError as e:
    print('Error:', e)
```

```python
#Final balance

print(ba.display_balance())
```
