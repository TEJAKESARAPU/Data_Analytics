```python
# 4. Write a program using bitwise NOT (~) on a number and explain the result.

x = int(input('Enter x value:'))
result = ~x
print('Result of x~ is:', result)

''' Explaination: In python the not(~) operator flips all the bits of the number. The formula for (~) operator is '-(n+1)'. So here n = 5,
then the result will be x = -(5+1) = 6 '''
```

```python
# 7. Combine multiple bitwise operations in one expression and print result.

x =int(input('Enter x value:'))
y =int(input('Enter y value:'))

result = (x&y) | (x^y) & (~x) >> (x) << (y)

print('Result of the combined expression is:', result)
```
