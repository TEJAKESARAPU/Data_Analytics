```python
# Q2. Count and print the total number of words in the article.

with open('article.txt', 'r') as f:
    text = f.read()
    words = text.split()
    print(f'Total no.of words: {len(words)}')
```

```python
# Q3. Count how many times each vowel appears in the article.

with open('article.txt', 'r') as f:
    text = f.read().lower()
    vowels = 'aeiou'
    vowel_count = {v: 0 for v in vowels}  #runs thru each character in the vowels

    for char in text:
        if char in vowels:
            vowel_count[char] += 1

    for v,c in vowel_count.items():
        print(f'{v}:{c}')
```

```python
# Q4. Find the longest word in the article.

with open('article.txt', 'r') as f:
    text = f.read()
    word = text.split()
    max_word = max(word, key=len)
    print(f'longest word: {max_word}')
```

```python
# Q5. Print all sentences that contain a given keyword.

keyword = input('Enter a keyword:').lower()

with open('article.txt', 'r') as f:
    text = f.read()
    sentences = text.split('.')
    for sentence in sentences:
        if keyword in sentence.lower():
            print(sentence.strip())
```

```python
# Q6. Count the number of words that start with a given letter.

starts = input('Enter a letter:')

with open('article.txt', 'r') as f:
    text = f.read()
    words = text.split()
    count = 0
    for word in words:
        if word.startswith(starts):
            count +=1

    print(f'No.of words starting with {starts}: {count}')
```

```python
# Q12. Reverse the content of the file and write it into reverse.txt.

with open('article.txt', 'r') as f:
    text = f.read()
    rev = text[::-1]

with open('reverse.txt', 'w') as f:
    f.write(rev)

print('Reversed content has been written to reverse.txt')
```

```python
# Q14. Find how many unique words are in the article (ignore case).

with open('article.txt', 'r') as f:
    text = f.read().lower()
    words = text.split()
    unique_words = set(words)
    print(f'No.of unique words: {len(unique_words)}')
```

```python
# Q15. Count how many times each word appears and save the result into word_count.txt.

with open('article.txt', 'r') as f:
    text = f.read().lower()
    words = text.split()
    word_count = {}
    
    for word in words:
        if word in word_count:
            word_count[word]+=1
        else:
            word_count[word] = 1

with open('word_count.txt', 'w') as f:
    for w, c in word_count.items():
        f.write(f'{w}: {c}\n')

print('Word counts has been saved into word_count.txt')
```

```python
# Q1. Write a Python program to read the last n lines of a file.

n = int(input('Enter number of lines to read from the end:'))

with open('article2.txt', 'r', encoding="utf-8") as f:
    lines = f.readlines()
    last_lines = lines[-n:]

    for line in last_lines:
        print(line.strip())
```

```python
# Q2. Write a Python program to count the frequency of words in a file.

with open('article2.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()
    words = text.split()
    frequency = {}

    for word in words:
        if word in frequency:
            frequency[word] +=1
        else:
            frequency[word] = 1

    for w, c in frequency.items():
        print(f'{w}: {c}')
```

```python
# Q2. Write a Python program to count the frequency of words in a file.

from collections import Counter
import re

with open('article2.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()
    words = re.findall(r'\b\w+\b', text)
    frequency = Counter(words)

    for w, c in frequency.items():
        print(f'{w}: {c}')
```

```python
# Q4. Write a Python program to combine each line from first file with the corresponding line in second file.

with open('article.txt', 'r', encoding='utf-8') as f1, open('article2.txt', 'r', encoding='utf-8') as f2:
    lines1 = f1.readlines()
    lines2 = f2.readlines()
    
with open('combined.txt', 'w') as out:
    for l1, l2 in zip(lines1, lines2):
        out.write(l1.strip() + ' ' + l2.strip() + '\n')

print("Lines combined and saved into combined.txt")
```

```python
# Q5.Write a Python program to write 10 random numbers into a file. 
# Read the file and then sort the numbers and then store it to another file.

import random

with open('numbers.txt', 'w') as f:
    for _ in range(10):
        num = random.randint(1,100)
        f.write(str(num)+'\n')
        
print('10 random numbers written to numbers.txt')

with open('numbers.txt', 'r') as f:
    nums = [int(x) for x in f]

with open('sorted_numbers.txt', 'w') as f:
    for n in sorted(nums):
        f.write(str(n) + '\n')

print("Sorted numbers written to sorted_numbers.txt")

with open('sorted_numbers.txt', 'r') as f:
    print(f.read())
```
