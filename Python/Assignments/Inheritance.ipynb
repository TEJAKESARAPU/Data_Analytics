```python
# Q2. Define a class Employee with a method details(). Inherit it in the class Manager and add a method assign_task(). Use a Manager object to call both methods.

class Employee:
    def __init__(self, name, emp_id, role):
        self.name = name
        self.emp_id = emp_id
        self.role = role
        
    def details(self):
        return f'Name: {self.name}, Employee ID: {self.emp_id}, Role: {self.role}'

class Manager(Employee):
    def assign_task(self):
        return f'{self.name} is assigning tasks to the Team.'
        
mgr = Manager('Teja', 102983101, 'Analyst Manager')
emp1 = Employee('Ram', 102983106, 'Data Analyst')
emp2 = Employee('Sarayu', 109871025, 'Support Team')

print(mgr.details())

employees = [emp1, emp2]

for emp in employees:
    print(mgr.assign_task(), '-->', emp.details())
```

```python
# Q3. Write a program where class Account has a method deposit(). Inherit it in class SavingsAccount and add a method add_interest(). 
# Demonstrate both methods using a SavingsAccount object.

class Account:
    def __init__(self, name, initial_balance=0):
        self.name = name
        self.balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f'Deposited Amount: {amount}, Current Balance: {self.balance}'
        else:
            return f'Deposit amount must be Positive'

class SavingsAccount(Account):
    def add_interest(self, rate):
        interest = self.balance * (rate/100)
        self.balance += interest
        return f'Interest Added: {interest}, New Balance: {self.balance}'

s1= SavingsAccount('Teja', 10000)
print(s1.deposit(500))
print(s1.add_interest(5))
```

```python
# Q4. Create a class Father with a method work() and a class Mother with a method care(). 
# Inherit both into class Child and call both methods using a Child object.

class Father:
    def work(self):
        return f'Father is the family head'

class Mother:
    def care(self):
        return f'Mother cares more for her children'

class Child(Father, Mother):
    def play(self):
        return f'Children love their parents'

s = Child()
print(s.work())
print(s.care())
print(s.play())
```

```python
# Q5. Define class Teacher with method teach(), and class Mentor with method guide(). 
# Inherit both in class Trainer. Create a Trainer object and call both methods.

class Teacher:
    def teach(self):
        return f'Teacher is someone who teaches you the knowledge'

class Mentor:
    def guide(self):
        return f'Mentor is someone who guides you to your success'

class Trainer(Teacher, Mentor):
    def support(self):
        return f'Trainer is both a teacher and a mentor'

st = Trainer()
print(st.teach())
print(st.guide())
print(st.support())
```

```python
# Q6. Create classes A and B, each with methods showA() and showB() respectively. Inherit both in class C. 
# Demonstrate method resolution order (MRO) by calling both methods using a C object.

class A:
    def showA(self):
        return f'This is A'

class B:
    def showB(self):
        return f'This is B'

class C(A,B):
    def showC(self):
        return f'This is C'

test = C()
print(test.showA())
print(test.showB())
print(test.showC())
```

```python
# Q7. Create a class Person with a method show_name(). Inherit class Student from Person, and then class CollegeStudent from Student.
# Call the method chain using a CollegeStudent object.

class Person:
    def show_name(self):
        return f'This is a person'

class Student(Person):
    def study(self):
        return f'Student studies'

class CollegeStudent(Student):
    def learn(self):
        return f'College students participate in the sports days'

test = CollegeStudent()
print(test.show_name())
print(test.study())
print(test.learn())
```

```python
# Q8. Write a program with a class Vehicle having a method move(). 
# Inherit class Car from Vehicle, and further inherit class ElectricCar from Car. 
# Demonstrate calling methods from all levels using an ElectricCar object.

class Vehicle:
    def move(self):
        return 'Vehicle moves on the tyres'

class Car(Vehicle):
    def diesel(self):
        return 'Car runs on Diesel'

class ElectricCar(Car):
    def current(self):
        return 'Electric car runs on Electric power'

car1 = ElectricCar()
print(car1.move())
print(car1.diesel())
print(car1.current())
```

```python
# Q10. Create a base class Shape with a method area(). 
# Inherit two classes, Square and Circle, from it. Implement area calculation in both and demonstrate with objects.

import math

class Shape:
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return f'Area of Square: {self.side * self.side}'

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return f'Area of Circle: {math.pi * self.radius * self.radius}'
sq = Square(5)
print(sq.area())
ci = Circle(8)
print(ci.area())
```

```python
# Q11. Write a program where the class Employee is the parent. Inherit classes Developer and Tester from it. 
# Each subclass should have its own method. Call all methods using their objects.

class Employee:
    def work(self):
        return f'Employee works for the firm'

class Developer(Employee):
    def develop(self):
        return f'Developer writes code'

class Tester(Employee):
    def test(self):
        return f'Tester run tests on the code'

emp = Employee()
dev = Developer()
tes = Tester()
print(emp.work())
print(dev.develop())
print(tes.test())
```

```python
# Q13. Create classes A (base), B and C that inherit from A, and class D inherits from both B and C. Demonstrate method calls to show hybrid inheritance.

class A:
    def classA(self):
        return 'This is class A'

class B(A):
    def classB(self):
        return 'This is class B'

class C(A):
    def classC(self):
        return 'This is class C'

class D(B, C):
    def classD(self):
        return 'This is class D'
clA = A()
clB = B()
clC = C()
clD = D()

print(clD.classA())
print(clD.classB())
print(clD.classC())
print(clD.classD())
```

```python
# Q14. Write a program where the class University is the parent. The College and Library are inherited from the University. 
# Another class, Student, inherits from both College and Library. Create a Student object and call methods from all classes.

class University:
    def department(self):
        return 'This is University'

class College(University):
    def branch(self):
        return 'This is college'

class Library(University):
    def book(self):
        return 'This is Library'

class Student(College, Library):
    def study(self):
        return 'This is Student'

st = Student()

print(st.department())
print(st.branch())
print(st.book())
print(st.study())
```

```python
# Q15. Define the class ElectronicDevice as the base. Create classes Computer and Mobile inheriting from it. 
# Then create the class SmartDevice inheriting from both Computer and Mobile. Show how methods from all classes are accessible.

class ElectronicDevice:
    def device(self):
        return 'This is an Electronic Device'

class Computer(ElectronicDevice):
    def laptop(self):
        return 'This is a Laptop'

class Mobile(ElectronicDevice):
    def call(self):
        return 'This is a Mobile'

class SmartDevice(Computer, Mobile):
    def automation(self):
        return 'This is a Smart Device'

sm = SmartDevice()

print(sm.device())
print(sm.laptop())
print(sm.call())
print(sm.automation())
```
