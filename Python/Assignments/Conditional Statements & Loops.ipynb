```python
# 1. Print numbers 1 to 5
# Use a while loop to print numbers from 1 to 5.

i = 1
while i <= 5:
    print(i)
    i = i+1
```

```python
# 3.Sum of natural numbers
# Find the sum of numbers from 1 to n using a while loop.

i = 1
n = int(input("Enter a number:"))
sum_of_numbers = 0

while i <= n:
    sum_of_numbers +=i
    i = i + 1
print(f'Sum of numbers from 1 and {n} is {sum_of_numbers}')
```

```python
# 4. Reverse digits
# Take a number as input and use a while loop to print its digits in reverse order.

n = int(input('Enter a number:'))

while n > 0:
    print(n % 10) #extracts the last digit
    n //=10 # removes the last digit
```

```python
# 1. Largest of Three Numbers
# Take three numbers as input and use if-elif-else to find and print the largest number.

a = int(input('Enter first number:'))
b = int(input('Enter second number:'))
c = int(input('Enter third number:'))

if a>= b and a>=c:
    print(f'{a} is the largest number')
elif b>=a and b>=c:
    print(f'{b} is the largest number')
else:
    print(f'{c} is the largest number')
```

```python
# 2. Check Even or Odd (with nested condition)
# Ask the user for a number and check if it’s even, print whether it’s also a multiple of 4. If it’s odd, print “Odd number”.

num = int(input('Enter a number'))

if num % 2 == 0:
    if num % 4 == 0:
        print(f'{num} is a even number and also a multiple of 4')
    else:
        print(f'{num} is a even number')
else:
    print(f'{num} is a odd number')
```

```python
# 3. Leap Year Checker
# Write a program to check whether a given year is a leap year or not. Hint: divisible by 4 but not by 100, unless also divisible by 400.

year = int(input('Enter a year:'))

''' condition - If the year is divisible by four, it's a leap year. 
But if the year can be divided by 100 as well as four, it's not a leap year. 
However, if the year is divisible by 400, it is a leap year. '''

if year % 4 == 0 or year % 100 == 0 and year % 400 == 0:
    print(f'{year} is a leap year')
else:
    print(f'{year} is not a leap year')
```

```python
# 4. Grade Calculator
# Input student marks and assign grades
# 90–100 → A
# 80–89 → B
# 70–79 → C
# 60–69 → D
# Below 60 → F

marks = int(input('Enter your marks:'))
if marks > 100 or marks < 0:
    print(f'Please enter valid marks (0-100)')
elif marks >= 90:
    print(f'You have secured Grade A')
elif marks >= 80:
    print(f'You have secured Grade B')
elif marks >= 70:
    print(f'You have secured Grade C')
elif marks >= 60:
    print(f'You have secured Grade D')
else:
    print(f'You have failed the test')
```

```python
# 5. Traffic Light System
# Ask the user for a traffic light color (red, yellow, or green) and display the appropriate message
# Red → Stop
# Yellow → Ready to go
# Green → Go


color = input('Please enter the traffic light color (red/yellow/green):').lower()

if color == 'red':
    print(f'Stop')
elif color == 'yellow':
    print(f'Ready to go')
elif color == 'green':
    print(f'Go')
else:
    print('Please enter valid traffic light color')
```

```python
# 7. Simple Calculator
# Take two numbers and an operator (+, -, *, /) as input, and perform the operation using if-elif.

a = float(input('Enter first number:'))
b = float(input('Enter second number:'))
operator = input("Enter operator (+, -, *, /): ")

if operator == '+':
    print(f'The result is {a+b}')
elif operator == '-':
    print(f'The result is {a-b}')
elif operator == '*':
    print(f'The result is {a*b}')
elif operator == '/':
    if b != 0:
        print(f'The result is {a/b}')
    else:
        print("Error: Division by zero is not allowed")
else:
    print('Invalid operator, enter a valid operator')
```

```python
# 8. Days in a Month
# Ask the user for the month number (1–12) and display how many days it has. Handle February (leap year check optional).

month = int(input('Enter a month number (1-12):'))
year = int(input('Enter the year:'))

if month > 12 or month < 1:
    print('Enter valid month number')
elif month == 2:
    if year % 4 == 0 or year % 100 == 0 and year % 400 == 0:
        print(f'29 days are present for the said month as the year {year} is a leap year')
    else:
        print(f'28 days are present for the said month')
elif month % 2 == 0:
    print(f'30 days are present for the said month')
else:
    print(f'31 days are present for the said month')
```

```python
# 9. Password Validation
# Ask the user for a password and check, 
# If the password length ≥ 8, 
# If it contains at least one number
# If it contains at least one special character
# Print messages accordingly. 

password = input("Enter your password: ")
special_characters = "!@#$%^&*()-_=+[]{};:'\",.<>?/\\|`~"
errors = []

# Check length
if len(password) < 8:
    errors.append("Password must be at least 8 characters long")

# Check number
if not any(ch.isdigit() for ch in password):
    errors.append("Password must contain at least one number")

# Check special character
if not any(ch in special_characters for ch in password):
    errors.append("Password must contain at least one special character")

if errors:
    for err in errors:
        print(err)
else:
    print("Password is valid ✅")
```
