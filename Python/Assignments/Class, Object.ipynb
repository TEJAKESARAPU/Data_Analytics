```python
# Q1. What is a class in Python? 

# Ans. A class in Python serves as a blueprint for objects,
# combining attributes and methods into one unit to model real-world entities in code.
```

```python
# # Q2. Explain what an object is in Python?

# # Ans. An object in Python is a real-world entity created from a class, holding its own data and 
# # capable of performing actions defined by that class. Objects are instances of classes.
# They encapsulate both data (attributes) and behaviour (methods).
```

```python
# Q3. How does class differ from an object? How do you create an object from a class?

# Ans. A class is a blueprint or template that defines properties and methods. An object is an instance of a class that holds actual data.
# Below is hoe to create an object from a class

class Student:     #class
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Creating objects
s1 = Student("Teja", 23)
s2 = Student("Anusha", 22)

print(s1.name, s1.age)
print(s2.name, s2.age)

# Here, 'Student' is the class, 's1, s2' are the objects
```

```python
# Q4. What are instance variables and class variables? Give an example of each.

# Instance variables are variables that belong to a specific instance (object) of a class. 
# Each object created from the class has its own independent copy of these variables. It is defined inside a constructor (__init__)
# Changing one object’s instance variable does not affect others.

# Class variables are the variables that belongs to the class itself, shared across all objects. Defined directly inside the class (but outside any methods).
# All instances share the same value unless explicitly overridden.

class Car:
    def __init__(self, make, model):
        self.make = make  # Instance variable
        self.model = model # Instance variable

car1 = Car("Toyota", "Camry")
car2 = Car("Honda", "Civic")

print(f"Car 1: {car1.make} {car1.model}")
print(f"Car 2: {car2.make} {car2.model}")

car1.make = "Ford" # Changing instance variable for car1
print(f"Car 1 (modified): {car1.make} {car1.model}") # Output: Car 1 (modified): Ford Camry
print(f"Car 2: {car2.make} {car2.model}") # Output: Car 2: Honda Civic (unaffected)
```

```python
# Q3. Add a method is_pass() that returns "Pass" if marks ≥ 40, otherwise "Fail".

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def is_pass(self):
        return f'{self.name} : {'Pass' if self.marks >= 40 else 'Fail'}'

stu1 = Student('Teja', 80)
stu2 = Student('Ram', 35)
print(stu1.is_pass())
print(stu2.is_pass())
```

```python
# Q4. Add a method grade() in the class that returns the grade based on marks:
# A → marks ≥ 90
# B → marks ≥ 75
# C → marks ≥ 60
# D → marks ≥ 40
# F → marks < 40

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def is_pass(self):
        return f'{self.name} : {'Pass' if self.marks >= 40 else 'Fail'}'

    def grade(self):
        if self.marks >= 90:
            return f'{self.name} : Grade A'
        elif self.marks >= 75:
            return f'{self.name} : Grade B'
        elif self.marks >= 60:
            return f'{self.name} : Grade C'
        elif self.marks >= 40:
            return f'{self.name} : Grade D'
        else:
            return f'{self.name} : Grade F'

stu1 = Student('Teja', 80)
stu2 = Student('Ram', 35)
stu3 = Student('Shyam', 95)
stu4 = Student('Tarzan', 65)
print(stu1.is_pass())
print(stu1.grade())
print(stu2.is_pass())
print(stu2.grade())
print(stu3.is_pass())
print(stu3.grade())
print(stu4.is_pass())
print(stu4.grade())
```
