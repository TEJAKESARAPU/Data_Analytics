```python
# 7. Write a Python program where you will return a dictionary where the keys will be the character and the values will be the occurrence of the 
# character in the string, and it should for any string inputs (dynamic) 
# Input: “my name is ame” 
# Output: {“m”:3,” y”:1,” n”:1,” a”:2,” e”:2,” i”:1,” s”:1}

str1 = 'my name is ame'
chr_count = {}

for ch in str1.replace(' ', ''):
    chr_count[ch] = chr_count.get(ch, 0) + 1

print(chr_count)
```

```python
# 9. Write a Python program to create a Rock, Paper, Scissors game, use two inputs for two players: 
# Conditions: -Rock beats Scissors, Scissors beats Paper, Paper beats Rock 

a = input('Enter Player A input (rock/paper/scissors)').lower()
b = input('Enter Player B input (rock/paper/scissors)').lower()

if a == b:
    print('It\'s a tie!')
elif (a == 'rock' and b == 'scissors') or \
     (a == 'scissors' and b == 'paper') or \
     (a == 'paper' and b == 'rock'):
    print('Player A wins!')
else:
    print('Player B wins!')
```

```python
# 21. Write a Python program that repeatedly accepts integer inputs from the user.
# The program should stop taking inputs when the user enters ‘q’.
# After exiting, display the average and product of all the entered numbers.

nums= []

for i in range(10):
    x = input('Enter a number (q to quit):')
    if x == 'q':
        break
    nums.append(int(x))

if nums:
    print('Average:', sum(nums)/len(nums))
    p = 1
    for n in nums:
        p *= n
    print('Product:', p)
else:
    print('No numbers entered')
```

```python
# 28. Write a python program that takes a string, breaks it up and returns it with vowels first, consonants second. 
# For any character that's not a vowel (like special characters or spaces), treat them like consonants

str1 = input('Enter a string:')

vowels = ''
consonants = ''

for ch in str1:
    if ch.lower() in 'aeiou':
        vowels += ch
    else:
        consonants += ch

print('Result:', vowels + consonants)
```

```python
# 29. Create a dynamic calculator that asks for numbers and an operator and returns the answers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, //, %): ")

if op == '+':
    print('Result:', a+b)
elif op == '-':
    print('Result:', a-b)
elif op == '*':
    print('Result:', a*b)
elif op == '/':
    print('Result:', a/b)
elif op == '//':
    print('Result:', a//b)
elif op == '%':
    print('Result:', a%b)
else:
    print('Invalid operator')
```

```python
# # 31. Armstrong Number – Comprehension Task
# # Check whether a given number is an Armstrong number using list comprehension.
# A number is called an Armstrong number if the sum of each digit raised to the power of the number of digits equals the number itself.
#  Example:
#  # 153 → 1³ + 5³ + 3³ = 153


num = int(input("Enter a number: "))
power = len(str(num))

if num == sum([int(d) ** power for d in str(num)]):
    print('Armstrong number')
else:
    print('Not an Armstrong number')
```

```python
# 32. Factorial of a Number (Recursion)
# Write a recursive function to find the factorial of a number n.
# Input: 5
# Expected Output: 120

def factorial(n):
    if n == 0 or n ==1:
        return 1
    return n * factorial(n-1)

num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

n = int(input("Enter n:"))
print("Fibonacci number:", fibonacci(n))
```

```python
# 34. Sum of Digits of a Number (Recursion)
# Write a recursive function to find the sum of digits of a given number
# Input: 987
# Expected Output: 24 
# (9 + 8 + 7)

def sum_of_digits(n):
    if n == 0:
        return 0
    return(n%10) + sum_of_digits(n//10)

num = int(input('Enter a number:'))
print('Sum of digits:', sum_of_digits(num))
```

```python
# 35. Reverse a String using Recursion
# Write a recursive function to reverse a string.
# Input: "python"
# Expected Output: "nohtyp"

def reverse_string(s):
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

str1 = input('Enter a string:')
print('Reversed string:', reverse_string(str1))
```

```python
# 35. Power of a Number (xⁿ using Recursion)
# Write a recursive function to compute xⁿ (x raised to the power n).
# x = 2
# n = 10
# Expected Output: 1024


def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n - 1)

x = int(input("Enter base number:"))
num = int(input("Enter power:"))
print("x power n:", power(x, num))
```
