```python
# 1. Using map with a lambda function to square the numbers [1, 2, 3, 4], is the result [1, 4, 9, 16]?

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
# 2. If a filter with a lambda function keeps numbers greater than 5 from [3, 6, 1, 8, 4, 7], is the result [6, 8, 7]?

numbers = [3, 6, 1, 8, 4, 7]
result = list(filter(lambda x: x > 5, numbers))
print(result)
```

```python
# 3. If a lambda function adds 2 to each element in [10, 20, 30] using map, is the result [12, 22, 32]?

numbers = [10, 20, 30]
result = list(map(lambda x: x+2, numbers))
print(result)
```

```python
# 4. If filter with a lambda function selects even numbers from [1, 2, 3, 4, 5, 6], is the length of the result 3?

numbers = [1, 2, 3, 4, 5, 6]
result = list(filter(lambda x: x % 2 == 0, numbers))
print(len(result))
```

```python
# 5. Using map with a lambda function to compute the cube of [2, 3, 4], is the sum of the result 99?

numbers = [2,3,4]
cubes = list(map(lambda x: x**3, numbers))
result = sum(cubes)
print(result)
```

```python
# 6. If a lambda function doubles each element in [5, 10, 15] using map, is the result [10, 20, 30]?

numbers = [5, 10, 15]
double = list(map(lambda x: x*2, numbers))
print(double)
```

```python
# 7. Using filter with a lambda function to keep numbers divisible by 3 from [3, 4, 6, 7, 9], is the result [3, 6, 9]?

numbers = [3, 4, 6, 7, 9]
divisible_3 = list(filter(lambda x: x%3==0, numbers))
print(divisible_3)
```

```python
# 8. If a lambda function multiplies each number in [1, 2, 3] by 2 and then adds 1 using map, is the result [3, 5, 7]?

numbers = [1, 2, 3]
print(list(map(lambda x: (x*2)+1, numbers)))
```

```python
# 9. Using filter with a lambda function to select numbers less than 10 from [5, 12, 8, 15, 3], is the length of the result 3?

numbers = [5, 12, 8, 15, 3]
print(len(list(filter(lambda x: x<10, numbers))))
```

```python
# 10. If a lambda function computes x % 2 == 0 with filter on [2, 3, 4, 5, 6], is the sum of the result 12?

numbers = [2, 3, 4, 5, 6]
print(sum(list(filter(lambda x: x % 2 == 0, numbers))))
```
