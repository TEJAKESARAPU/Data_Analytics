```python
# 1. Factorial of a Number
# Write a program to find the factorial of a number using a for or while loop.
# Example: 5! = 120

num = int(input('Enter a number to find it\'s factorial:'))
factorial = 1

for i in range(1, num+1):
    factorial *= i

print(f'The factorial of {num} is {factorial}')
```

```python
# 2. Count Digits in a Number
# Use a while loop to count how many digits are in an integer.
# Example: 12345 → 5 digits

num = int(input('Enter a number:'))
count = 0

while num > 0:
    count +=1
    num //=10

print(f'Number of digits in are {count}')
```

```python
# 3. Reverse a String
# Take an input string and print it in reverse using a loop.
# Example: "Python" → "nohtyP"

str1 = input('Enter a string')
rev = ''

for ch in str1:
    rev = ch + rev
print(rev)
```

```python
# 5. Sum of Digits
# Take an integer as input and use a while loop to calculate the sum of its digits.
# Example: 123 → 1 + 2 + 3 = 6

num = int(input('Enter a number'))
sum_of_num = 0

while num > 0:
    sum_of_num += num % 10
    num=num//10
print(f'Sum of digits is {sum_of_num}')
```

```python
# 6. Fibonacci Series
# Generate and print the first n numbers of the Fibonacci sequence using a loop.
# Example: 0, 1, 1, 2, 3, 5, 8, …

n = int(input("Enter how many terms: "))

a, b = 0, 1
print(a,b, end=' ')

for i in range(2, n):
    c = a+b
    print(c, end=' ')
    a,b = b, c
```

```python
# 7. Find Prime Numbers in a Range
# Use a nested for loop to print all prime numbers between 1 and 50.

for num in range(2,51):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(f'{num} is a prime number')
```

```python
# 8. Count Vowels in a String
# Use a for loop to count the number of vowels (a, e, i, o, u) in a given string.

name = input('Enter a name:').lower()
vowels = 'aeiou'
vow = 0

for ch in name:
    for i in vowels:
        if ch == i:
            vow+=1
print((f'Number of vowels: {vow}'))
```

```python
# 9. Check for Palindrome
# Use a loop to check if a given string or number is a palindrome.
# Example: "madam" → Palindrome

text = input('Enter a name:').lower()
palindrome = ''

for ch in text:
    palindrome = ch + palindrome
    
if text == palindrome:
    print('Palndrome')
else:
    print('Not a palindrome')
```

```python
# 16. Count Even and Odd Digits
# Take an integer and count how many even and odd digits it has using a while loop.
# Example: Input: 12345 → Even=2, Odd=3

num = int(input('Enter a number:'))
even = 0
odd = 0

while num > 0:
    digit = num % 10
    if digit % 2 == 0:
        even +=1
    else:
        odd +=1
    num//=10

print(f'Even = {even} and odd = {odd}')
```

```python
# 17. Reverse a Number
# Use a while loop to reverse a given number.
# Example: Input: 1234 → Output: 4321

num = int(input('Enter a number:'))
rev = 0

while num > 0:
    i = num % 10
    rev = rev * 10 + i
    num //=10
print(rev)
```

```python
# 18. Skip Multiples of 3
# Print numbers from 1 to 20, but skip numbers that are multiples of 3 using continue.

for i in range(1,21):
    if i % 3 == 0:
        continue
    else:
        print(i)
```
