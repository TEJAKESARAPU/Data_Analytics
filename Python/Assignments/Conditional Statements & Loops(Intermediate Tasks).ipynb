```python
# 1. Factorial of a Number
# Write a program to find the factorial of a number using a for or while loop.
# Example: 5! = 120

num = int(input('Enter a number to find it\'s factorial:'))
factorial = 1

for i in range(1, num+1):
    factorial *= i

print(f'The factorial of {num} is {factorial}')
```

```python
# 2. Count Digits in a Number
# Use a while loop to count how many digits are in an integer.
# Example: 12345 â†’ 5 digits

num = int(input('Enter a number:'))
count = 0

while num > 0:
    count +=1
    num //=10

print(f'Number of digits in are {count}')
```

```python
# 3. Print Multiples of 3 and 5
# Use a for loop to print all numbers between 1 and 100 that are multiples of 3 or 5.

for i in range(1, 101):
    if i % 3 == 0 or i % 5 == 0:
        print(f'{i} is a multiple of 3 or 5')
```
