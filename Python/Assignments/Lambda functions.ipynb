```python
# Q1. Create a lambda function that converts temperature from Fahrenheit to Celsius.
#  Input: temp = 96.8
#  Output: 36

ftoc = lambda f: (f-32) * 5 / 9
temp = 96.8
print(int(ftoc(temp)))
```

```python
# Q2. Create a lambda function that finds the area of a circle.
#  Input: radius = 5
#  Output: 78.54

import math

area_circle = lambda r: math.pi * r**2
radius = 5
print(round(area_circle(5),2))
```

```python
# Q3. Create a lambda function that converts American dollars to Indian rupees.
#  Input: dollar_conversion = 45
#  Output: 3950

dollar_to_rupee = lambda dollars: dollars * 87.78
dollar_conversion = 45
print(round(dollar_to_rupee(dollar_conversion)))
```

```python
# Q4. Create a lambda function that identifies child and adult (adult ≥ 18).
#  Input: age = 10
#  Output: Child

identify_age = lambda age: 'Adult' if age >= 18 else 'Child'
age = int(input('Enter your age:'))
print(identify_age(age))
```

```python
# Q5. Create a lambda function that checks if a number is positive or negative.
#  Input: -5
#  Output: Negative

check_number = lambda num: 'Positive' if num > 0 else('Zer' if num == 0 else 'Negative')
num = -5
print(check_number(num))
```

```python
# Q6. Write a lambda function that checks if a number is even or odd.
#  Input: 7
#  Output: Odd

even_odd = lambda num: 'Even' if num % 2 == 0 else 'Odd'
num = 7
print(even_odd(num))
```

```python
# Q7. Create a lambda function that finds the maximum of two numbers.
#  Input: 10, 25
#  Output: 25

max_num = lambda num: max(num)
num = 10, 25
print(max_num(num))
```

```python
# Q8. Create a lambda function that checks if a person is eligible to vote (age ≥ 18).
#  Input: 16
#  Output: Not Eligible

eligible_to_vote = lambda age: 'Eligible' if age >= 18 else 'Not eligible'
age = 16
print(eligible_to_vote(age))
```

```python
# Q9. Write a lambda function that checks if a number is a multiple of 5.
#  Input: 20
#  Output: Yes

multiple_of_num = lambda num: 'Yes' if num % 5 == 0 else 'No'
num = 20
print(multiple_of_num(num))
```

```python
# Q10. Write a lambda function to determine if a character is a vowel or a consonant.
#  Input: 'e'
#  Output: Vowel

chr_in_vowels = lambda ch: 'Vowel' if ch.lower() in 'aeiou' else 'Consonant'
chr1 = 'e'
print(chr_in_vowels(chr1))
```

```python
# Q11. Create a lambda function that checks if a string is empty or not.
#  Input: ""
#  Output: Empty

empty_cell = lambda x: 'Empty' if x=="" else 'Not empty'
x = ""
print(empty_cell(x))
```

```python
# Q12. Write a lambda function that checks if a number lies between 10 and 50 (inclusive).
#  Input: 35
#  Output: Yes

num_freq = lambda x: 'Yes' if x >= 10 and x<=50 else 'No'
num = 35
print(num_freq(num))
```

```python
# Q13. Create a lambda function that returns Pass if marks ≥ 40, otherwise Fail.
#  Input: 28
#  Output: Fail

pass_fail = lambda x: 'Pass' if x >= 40 else 'Fail'
marks = 28
print(pass_fail(marks))
```

```python
# Q14. Write a lambda function that checks if two numbers are equal or not.
#  Input: 15, 20
#  Output: Not Equal

num_equal = lambda x, y: 'Equal' if x == y else 'Not equal'
x,y  = 15, 20
print(num_equal(x,y))
```

```python
# Q15. Define a lambda function that converts odd numbers to even.
#  Input: 5
#  Output: 6

odd_to_even = lambda x: x+1 if x % 2 != 0 else x
num = 5
print(odd_to_even(num))
```
