```python
# Q2. Write a Python program to convert only the characters at odd indexes to lowercase.
# Example:
# Input: PYTHON
# Output: PyThOn 

text = input('Enter a name:')
result = ''

for i in range(len(text)):
    if i % 2 == 0:
        result += text[i].upper()
    else:
        result += text[i].lower()
print(result)
```

```python
# Q3. Write a Python program to convert only the characters at even indexes to lowercase.
# Example:
# Input: PYTHON
# Output: pYtHoN

text = input('Enter a name:')
result = ''

for i in range(len(text)):
    if i % 2 == 0:
        result += text[i].lower()
    else:
        result += text[i].upper()
print(result)
```

```python
# Q5. Write a Python program to convert only the characters at even indexes to uppercase.
# Example:
# Input: python
# Output: PyThOn.

text = input('Enter some text:')
result = ''

for ch in range(len(text)):
    if ch % 2 == 0:
        result += text[ch].upper()
    else:
        result += text[ch]
print(result)
```

```python
# Q7. Write a Python program to count how many times the character '@' occurs in a given string.

text = input('Enter your password:')
char = '@'
count = 0

for i in range(len(text)):
    if text[i] == char:
        count += 1
print(f'count of @: {count}')
```

```python
# Q8. Write a Python program to extract only the names from the following string.
# "name1@gmail.com, name2@gmail.com, name3@gmail.com"
# Output: name1, name2, name3

text = "name1@gmail.com, name2@gmail.com, name3@gmail.com"
names = []

for ch in text.split(','):
    names.append(ch.split('@')[0].strip())
print(", ".join(names))
```
