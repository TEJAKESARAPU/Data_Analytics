```python
# Q6. Assign different data types
# Assign the following to variables and print their types:
# An integer
# A float
# A string
# A boolean

integer_ = int(input('Integer:'))
float_ = float(input('Float:'))
string_ = str(input('String:'))
bool_input = input('Boolean (True/False):')
boolean_ = bool_input.lower() == 'true'

print("Integer:", integer_, '|' ,'Type:', type(integer_))
print("Float:", float_, '|' ,'Type:', type(float_))
print("String:", string_, '|' ,'Type:', type(string_))
print('Boolean', boolean_, '|' ,'Type:', type(boolean_))
```

```python
# Q7. Dynamic typing
# Assign an integer to a variable num. Then reassign a string to the same variable. Print the value and type each time.

#assigning an integer
num = 55
print('Value of num is', num)
print('Type of num is', type(num))

#Re-assigning a string to the variable num
num = 'Fifty five'
print('Value of num is', num)
print('Type of num is', type(num))
```

```python
# Q8. Create a variable city and assign it the value "Bangalore". Then create another variable population with a value of 12000000. Print both variables in a single line separated by a comma.
# Input:
# city = "Bangalore"
# population = 12000000
# Output:
# Bangalore, 12000000

city = 'Bangalore'
population = 12000000
print(city + ',', population)
```

```python
# Q9. Create a variable language = "Python" and another variable version = 3.10. Print a sentence combining both variables like: "Python version is 3.10".
# Input:
# language = "Python"
# version = 3.10
# Output:
# Python version is 3.10

language = "Python"
version = 3.10
print(f"{language} version is {version:.2f}")
```

```python
# Q10. Concatenate variables
# Create two string variables first_name = "John" and last_name = "Doe". Concatenate them into a new variable, full_name and print it.

first_name = "John"
last_name = 'Doe'
full_name = first_name + ' ' + last_name
print(full_name)
```
