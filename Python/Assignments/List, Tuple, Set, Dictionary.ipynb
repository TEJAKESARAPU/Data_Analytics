```python
# Q1. Find common elements from three sets. 
# Input: {1,2,3,4}, {2,3,5,6}, {0,2,3,7} 
# Output: {2, 3}

set1, set2, set3 = {1,2,3,4}, {2,3,5,6}, {0,2,3,7}
print(set1 & set2 & set3)
```

```python
# Q2. Find elements that are only in the first set but not in the other 
# two. 
# Input: {1,2,3,4,5}, {2,3}, {4,6} 
# Output: {1, 5} 

set1, set2, set3 = {1,2,3,4,5}, {2,3}, {4,6}
print(set1 - set2 - set3)
```

```python
# Q3. Create a set of unique vowels from a string. 
# Input: "education" 
# Output: {'a', 'e', 'i', 'o', 'u'} 

num = "education"
vowels = 'aeiou'
result = set()

for ch in num:
    if ch in vowels:
        result.add(ch)
print(result)
```

```python
# Q4. Convert a list of tuples into a set. 
# Input: [(1,2), (3,4), (1,2)]
# Output: {(1, 2), (3, 4)} 

num = [(1,2), (3,4), (1,2)]
result = set(num)
print(result)
```

```python
# Q5. Find the intersection of characters in two words. 
# Input: "banana", "bandana" 
# Output: {'a', 'b', 'n'} 

fruit1, fruit2 = "banana", "bandana"
result = set(fruit1) & set(fruit2)
result
```

```python
# Q6. Find digits in the first number but not in the second. 
# Input: 12345, 34567 
# Output: {'1', '2'}

num1, num2 = '12345', '34567'
result = set(num1) - set(num2)
result
```

```python
# Q7. Find all unique words from a sentence. 
# Input: "python is fun and python is powerful" 
# Output: {'is', 'python', 'fun', 'and', 'powerful'}

sent = "python is fun and python is powerful"
unique_words = set(sent.split())
print(unique_words)
```

```python
# Q8. Remove duplicates from a list. 
# Input: [1,2,2,3,3,4,5] 
# Output: {1, 2, 3, 4, 5} 

num = [1,2,2,3,3,4,5]
print(set(num))
```

```python
# Q9. Find non-overlapping characters between two words. 
# Input: "hello", "world" 
# Output: {'h', 'e', ‘w’, 'r', 'd'} 

word1, word2 = "hello", "world"
result = set(word1) ^ set(word2)
print(result)
```

```python
# Q10. Find maximum and minimum elements in a set. 
# Input: {12, 7, 25, 3, 18} 
# Output: 25 3 

num = {12, 7, 25, 3, 18}
print(max(num), min(num))
```

```python
# 1) Create and print dictionary 
# Create a dictionary with 3 key–value pairs: name, age, and city. 
# Print all keys and values. 

dict1 = {
            'name' : 'Teja',
            'age' : '24',
            'city' : 'Visakhapatnam'
}

for key, value in dict1.items():
    print(key,':', value)
```

```python
# 2) Get the value of a dictionary with the maximum value. 
# Input: {'a':10, 'b':25, 'c':15} 
# Output: 25 

dict1 = {'a':10, 'b':25, 'c':15}
result = max(dict1.values())
result
```

```python
# 3) Get all dictionary values as a single string. 
# Input: {'a':'hi','b':'there'} 
# Output: "hi there" 

dict1 = {'a':'hi','b':'there'}
result = " ".join(dict1.values())
result
```

```python
# 4) Remove the last inserted pair. 
# Input: {'a':1, 'b':2, 'c':3} 
# Output: {'a':1, 'b':2}

dict1 = {'a':1, 'b':2, 'c':3}
dict1.popitem()
print(dict1)
```

```python
# 5) Find values present in both dictionaries. 
# Input: {'x':1, 'y':2}, {'a':2, 'b':3} 
# Output: {2} 

dict1 = {'x':1, 'y':2}
dict2 = {'a':2, 'b':3}
result = set(dict1.values()) & set(dict2.values())
print(result)
```

```python
# 6) Find the sum of dictionary keys. 
# Input: {10:'a', 20:'b', 30:'c'} 
# Output: 60

dict1 = {10:'a', 20:'b', 30:'c'}
print(sum(dict1.keys()))
```

```python
# 7) Remove duplicates from dictionary values. 
# Input: {'a':1, 'b':2, 'c':1} 
# Output: {1, 2} 

dict1 = {'a':1, 'b':2, 'c':1}
print(set(dict1.values()))
```

```python
# 8) Create a dictionary with the keys 'name', 'age', and 'city'. 
# Then: 
# a. Print all keys and values. 
# b. Access and print the value of the key 'course' from the dictionary student = {'name': 'Ravi', 'age': 20, 'course': 'Python'}. 
# c. Add a new key 'grade': 'A' to the dictionary and remove the key 'age'. 
# Finally, print the updated dictionary.

dict1 = {'name' : 'Teja', 'age': 24, 'city': 'Visakhapatnam'}
student = {'name': 'Ravi', 'age': 20, 'course': 'Python'}

# print all the key values and keys
for key, value in dict1.items():
    print(key, ':', value)

# Access and print the value of the key 'course' from the dictionary student
print('course:',student['course'])

# Add a new key 'grade': 'A' to the dictionary and remove the key 'age' and printing the updated dictionary
dict1['grade'] = 'A'
dict1.pop('age')
print(dict1)
```
