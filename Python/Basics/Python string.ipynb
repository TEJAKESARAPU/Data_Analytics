```python
name = 'rohan'
age = 25

print('Positional formatting: My name is {} and my age is {}'.format(name, age)) #positional formatting
print('Indexed formatting: My name is {1} and age is {0}'.format(age, name)) #indexed formatting
print('Named formatting: My name is {n} and my age is {a}'.format(n = 'rohan', a = 25)) #Names formatting
print(f'f-string formatting: My name is {name} an my age is {age}') #f-string formatting
```

```python
#count of upper, lower, spaces, capitalized, digit

#My laptop MODEl is m4.0      Pro

text = "My laptop MODEl is 'M4'     Pro"

upper = lower = digit = space = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
    elif ch.isspace():
        space += 1
    elif ch.isdigit():
        digit += 1

print('Upper:', upper)
print('Lower:', lower)
print('Space:', space)
print('digit:', digit)
```

```python
#palindrome

#take input from the user, we have to reverse the word and print whether the word is a palindrome or not using an if condition.

text = input('Enter a palindrome word:').lower()
rev = text[::-1]

if text == rev:
    print(f'{text} is a palindrome')
else:
    print(f'{text} not a palindrome')
```

```python
text = input('Enter a palindrome word:').lower()

rev = ""

for ch in text:
    rev = ch + rev

if rev == text:
    print(f'{text} is a palindrome')
else:
    print(f'{text} is not a palindrome')
```
