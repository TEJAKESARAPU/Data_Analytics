```python
# A bank account - each customer has a different balance, so deposits/withdrawals must
# work on the individual account

#class - bank account
#attributes - account holder name, balance
#method for deposit and display
#define object and call
#Instance method

class bank_account:
    bank_name = 'HDFC'
    
    def __init__(self, account_holder_name, balance):
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return f'{self.account_holder_name} deposited Rs.{amount}. New balance: Rs.{self.balance}'
    
    def display(self):
        return f'{self.account_holder_name} has {self.balance} rupees'

account1 = bank_account('Teja', 50000)
account2 = bank_account('Shanmukh', 100000)
account3 = bank_account('Dilip', 45000)
account4 = bank_account('Sam', 6000)

print(account1.deposit(5000))
print(account1.display())
print(account2.deposit(200))
print(account2.display())
```

```python
#class method

class student:
    school_name = 'DPS'

    def __init__(self, name, rollno):
        self.name = name
        self.rollno = rollno
    
    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name

    @classmethod
    def student_info(cls, stu_info):
        name, rollno = stu_info.split('-')
        return cls(name, rollno)

student.change_school('Innomatics')
stu1 = student.student_info('Mohit-01')
print(stu1.name, stu1.rollno)
print(student.school_name)
```

```python
#Static method

class bank:
    @staticmethod
    def is_valid_amount(amount):
        return amount > 0

print(bank.is_valid_amount(100))
print(bank.is_valid_amount(-12))
```

```python
#Static method

class addition:
    @staticmethod
    def add_num(a, b):
        return a+b

print(bank.add_num(100, 5))
```

```python
#Dunder/magic method

class student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __str__(self):
        return f'{self.name} scored {self.marks}'

    def __gt__(self, other):
        return self.marks > other.marks

stu1 = student('Teja', 95)
stu2 = student('Tara', 85)

print(stu1)
print(stu1 > stu2)
```

```python
#Property method

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary   # this will call the setter

    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError('Salary can not be negative')
        self._salary = value


# Test
emp = Employee('Rakesh', 60000)
print(emp.salary)

emp.salary = 70000
print(emp.salary)
emp.salary = -150
```

```python
#All togerther

class car:
    total_cars = 0
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    #Instance method
    def details(self):
        return f"My car is {self.brand} and the top speed is {self.speed}km/h"

    #class method
    @classmethod
    def show_total_cars(cls):
        return f'Total cars created: {cls.total_cars}'

    #static method
    @staticmethod
    def is_valid_speed(speed):
        return speed > 0

    #dunder/magic method
    def __str__(self):
        return f'{self.brand} car'

    #property method
    @property
    def speed_value(self):
        return self.speed

    @speed_value.setter
    def speed_value(self, new_speed):
        if new_speed < 0:
            raise ValueError('Speed cannot be negative')
        self.speed = new_speed

car1 = car('BMW', 200)
print(car1.details())

print(car.total_cars) #print class method
print(car.show_total_cars())  # Class method
print(car.is_valid_speed(150))  # Static method
print(car1) # Dunder method
print(car1.speed_value)  # Property method
car1.speed_value = 220
print(car1.speed_value)
```
