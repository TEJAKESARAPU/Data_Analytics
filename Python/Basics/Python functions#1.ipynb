```python
# Even or odd number

def num(n):
    print('Even' if n % 2 == 0  else 'Odd')
```

```python
num(4)
```

```python
# Define a function to check if a number is even or odd

def even_odd(num):
    if num % 2 == 0:
        return 'Even Number'
    else:
        return 'Odd Number'
```

```python
even_odd(3)
```

```python
# take the price of an item and add GST price to it
# GST price

def gst(price):
    return price + (0.18 * price)
```

```python
gst(100)
```

```python
#ATM Pin

def atm(pin):
    if pin == 1234:
        return 'Access granted!'
    else:
        return 'Wrong pin'
```

```python
atm(1234)
```

```python
atm(123)
```

```python
atm(4100)
```

```python
# Positional Arguments

def person(name, age):
    print(f'My name is {name} and my age is {age}')
```

```python
person('rohan', 26)
```

```python
person (26, 'rohan')
```

```python
def atm(bank, location):
    return f'{bank} atm is locatd at {location}'
```

```python
atm('HDFC', 'KPHB')
```

```python
atm('Kukatpally', 'SBI')
```

```python
''' Keyword Argument '''
```

```python
def atm(bank, position):
    return f'{bank} is located at {position}'
```

```python
atm(bank = 'SBI', position='Madhapur')
```

```python
atm(position='Madhapur', bank='HDFC')
```
