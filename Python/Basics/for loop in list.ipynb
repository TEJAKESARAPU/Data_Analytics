```python
fruits = ['apple', 'banana', 'mango']
```

```python
for fruit in fruits:
    print(fruit)
```

```python
for fruit in range(len(fruits)):
    print(f'{fruit} : {fruits[fruit]}') #here 'f' means formatting
```

```python
for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
# Real-life examples

marks = [40,50,60,70,55,75] 
passed_student = []

for mark in marks:
    if mark > 50:
        passed_student.append(mark)

passed_student
```

```python
numbers = [1,2,3,4,5,6,7,8]

even = []

for i in numbers:
    if i % 2 == 0:
        even.append(i)

even
```
