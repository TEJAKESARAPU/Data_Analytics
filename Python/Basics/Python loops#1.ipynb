```python
class laptop:
    def __init__(self, brand, price, color, ram):
        self.brand = brand
        self.price = price
        self.color = color
        self.ram = ram
lap1 = laptop('Acer', '100000', 'black', 16)
lap2 = laptop('Dell', 70000, 'silver', 8)

print(lap1.brand, lap1.color)
print(lap2.ram)
```

```python
#attributes/variables

class laptop:
    shop_name = 'Bajaj Electronics'    #class variables
    def __init__(self, brand, price, color, ram):   #instance variables
        self.brand = brand
        self.price = price
        self.color = color
        self.ram = ram
lap1 = laptop('Acer', '100000', 'black', 16)
lap2 = laptop('Dell', 70000, 'silver', 8)

print(lap1.brand, lap1.color)
print(lap2.ram)

print(lap2.shop_name)
print(lap2.shop_name)
```

```python
class Car:
    def __init__(self, make, model, year, color, transmission):
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.transmission = transmission

    def functionality(self):
        return f'My car is {self.make} {self.model} {self.year} year made. The color of the car is {self.color} and it\'s a {self.transmission} transmission car.'

car1 = Car('Maruthi Suzuki', 'VXI', 2013, 'pearl blue', 'manual')
car2 = Car('Hyundai', 'N-line', 2024, 'black', 'CVT')
```

```python
print(car1.functionality())
print(car2.functionality())
print(car1.make, car1.model)
```

```python
class phone:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        print('Brand:', self.brand)

ph1 = phone('Oneplus')
ph2 = phone('iPhone')

ph1.show_brand()
ph2.show_brand()
```

```python
phone.show_brand(ph2)
```

```python
# class attribute
# instance attribute
# constructor
# methods
# objects
# calculations

class products:
    website_name = 'Flipkart'
    def __init__(self, product_name, price, quantity):
        self.product_name = product_name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return f'Product name : {self.product_name}, price : {self.price}, quantity : {self.quantity}, total price : {self.price*self.quantity}'

product1 = products('Toy car', 500, 3)
product2 = products('books', 15, 8)
product3 = products('pen', 3, 5 )

print(product1.total_price())
print(product2.total_price())
print(product3.total_price())
print('Website:', products.website_name)
```
