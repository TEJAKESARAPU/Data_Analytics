```python
# laptop store and we have 3-4 things to purchase laptop, keyboard and mouse. To check if quantity is less than 6, we have to print that the item is out of stock 

store = {'laptop' : 6,
         'mouse' : 20,
         'keyboard' : 4}
for key, value in store.items():
    if value < 6:
        print(f'{key} is out of stock')
    else:
        print(f'{key} is available')
```
