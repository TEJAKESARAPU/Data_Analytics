```python
#Arithmetic Operator

# a=6
# b=2

a,b = 6, 2
```

```python
print('Addition:',a+b)
print('Substraction:',a-b)
print('Multiplication:',a*b)
print('Division:',a/b)
print('Floor Division:',a//b)
print('Modulus:',a%b)
print('Power:',a**b)
```
