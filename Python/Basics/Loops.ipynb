```python
#While loop

i = 0
while i<=5:
    print(i)
    i = i+1
```

```python
#for loop

for i in range(5):
    print(i)
```

```python
i = 1
while i <=5:
    print(f'You are at floor: {i}')
    i +=1
```

```python
# break condition

i = 1
while i<=5:
  if i == 3:
    break
  print(i)
  i = i + 1
```

```python
# continue condition

i = 0
while i <=5:
  if i == 3:
    i = i + 1
    continue
  print(i)
  i = i + 1
```

```python
i = 0
while i <=5:
  i +=1
  if i == 3:
    continue
  print(i)
```

```python
# else with a while loop

i = 1
while i <=3:
  print('Hi', i)
  i +=1
else:
  print('Loop is completed')
```

```python
#sum of number

i = 1
sum = 0
while i <=10:
  sum += i
  i+=1
print(sum)
```

```python
#count the digit

num = int(input('Enter a number: '))
count = 0
while num > 0:
  count +=1
  num //= 10

print(f'Number of digits are {count}')
```
