```python
gst = 0.18   #Global variable
def total_bill():
    price = 1500   #Local variable
    total = price + (price * gst)
    print(f'customer bill: {total}')
    
total_bill()
print(f'Current GST: {gst}')
```

```python
student = 'teja'
course = 'data_analytics'

def student_detais():
    age = 24
    roll = 101
    print('Local namespace:', locals())

# Show all global names
print("Global namespace keys:")
print(list(globals().keys()))

# Show specific global variables
print("Global variables:")
print({k: globals()[k] for k in ['student', 'course', 'student_detais']})
```

```python
def search(folder):
    for item in folder:
        if isinstance(item, list):
            search(item)    #recursive
        else:
            print('file', item)

data = ['file1.pdf', ['photos', ['image.png', 'image2.png']], 'notes.txt']
search(data)
```

```python
def calculate_interest(principal, rate, months):
    if months == 0:
        return 0
    interest = principal * rate
    return interest + calculate_interest(principal, rate, months -1)

print('Total interest', calculate_interest(10000, 0.06, 24))
```

```python
# fibonacci series

num = int(input('Enter a number:'))

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
print('Fibonacci Series:')

for i in range(num):
    print(str(fibonacci(i)) + ' ', end='')
```
