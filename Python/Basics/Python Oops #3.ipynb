```python
class animal:    #parent class
    def sound(self):
        print('This animal makes some sound')

class dog(animal):    #child class
    pass

dog1 = dog()
dog1.sound()
```

```python
#method overriding

class animal:    #parent class
    def sound(self):
        print('This animal makes some sound')

class dog(animal):  #child class
    def sound(self):
        print('Dog\'s sound')     #overriden

dog1 = dog()
dog1.sound()
```

```python
#using constructor with super()

class person:
    def __init__(self, name):
        self.name = name
class student(person):
    def __init__(self, name, roll):
        super().__init__(name)      #calling parent constructor using super() function
        self.roll = roll

stu1 = student('Mohan', 21)
print(stu1.name, stu1.roll)
```

```python
# create a class for a car, create a child class and inherit the properties from the car class

class car:     #parent class
    def __init__(self, name, model):   #using constructor
        self.name = name
        self.model = model
class version(car):    #child class
    def __init__(self, name, model, varient):
        super().__init__(name, model)       #calling parent constructor using super() function
        self.varient = varient

car1 = version('Audi', 'Q4', 'mid-version')
car2 = version('Benz', 'C-class', 'Top-end')

print(car1.name, car1.model, car1.varient)
print(car2.name, car2.model, car2.varient)
```

```python
class vehicle:
    def start(self):   #using method
        print('Vehicle is ready to start')

class bike(vehicle):
    def start(self):
        super().start()
        print('Now you can ride the vehicle')
bike1 = bike()
bike1.start()
```

```python
#Single inheritance
#one parent - one child class
class A:
    def show(self):
        print('A')
        
class B(A):
    pass
    
B().show()
```

```python
#Multi-level inheritance
#grandparent, parent, child class

class A:
    def showA(self):
        print('A')

class B(A):
    def showB(self):
        print('B')

class C(B):
    def showC(self):
        print('C')

obj = C()
obj.showA()
obj.showB()
obj.showC()
print()
C().showC()
C().showB()
C().showA()
```

```python
#multiple inheritance
#two parent classes but one child class

class A:                  #parent class
    def showA(self):
        print('A')
class B:                  #parent class
    def showB(self):
        print('B')
        
class C(A,B):             #class class
    pass

obj = C()
obj.showA()
obj.showB()
```

```python
#Hierarchical inheritance

class A:
    def show(self):
        print('A')

class B(A):
    pass

class C(A):
    pass

B().show()
C().show()
```

```python
# create a class for a vehicle and inherit the property in multiple child classes

class vehicle:
    def __init__(self, name):
        self.name = name
        print(f'Vehicle name is {self.name}')

class car(vehicle):
    pass

class bike(vehicle):
    pass

car1 = car('Creta')
bike1 = bike('Yamaha MT-15')
```

```python
from abc import ABC, abstractmethod

class payment(ABC):                        #abstract class     (Abstraction)
    def __init__(self, amount):
        self._amount = amount              #protected variable (Enapsulation)

    @abstractmethod
    def process_payment(self):
        pass

class ccpayment(payment):                  #child class (Inheritance)
    def __init__(self, amount, card_number):
        super().__init__(amount)
        self.card_number = card_number

    def process_payment(self):
        print(f'Processing credit card payment of {self._amount} using card number {self.card_number}')

class upi_payment(payment):
    def __init__(self, amount, upi_id):
        super().__init__(amount)
        self.upi_id = upi_id

    def process_payment(self):
        print(f'Processing UPI payment of {self._amount} using UPI ID {self.upi_id}')

class order:
    def __init__(self, item_name, price, quantity):
        self.item_name = item_name
        self.price = price
        self.quantity = quantity

    def get_total(self):
        return self.price * self.quantity

class user:
    def __init__(self, user_name, address):
        self.user_name = user_name
        self.address = address

    def get_details(self):
        return f'Customer : {self.user_name}, Address : {self.address}'

user1 = user('Teja', 'Visakhapatnam')
print(user1.get_details())
order1 = order('Laptop', 99990, 2)
total_amount = order1.get_total()
print('Total amount:',total_amount)
payment_method = ccpayment(total_amount, '1234-4567-7890')
payment_method.process_payment()
```
