```python
#try -except
#zero division error

try:
    x = 10/0
except ZeroDivisionError:
    print('Cannot divide by zero!')
```

```python
#multiple except block

try:
    num = int(input('Enter a number:'))
    print(10/num)
    
except ZeroDivisionError:
    print('Number cannot be zero!')

except ValueError:
    print('Please enter numbers!')
```

```python
try:
    n = int(input('Enter your age:'))
except ValueError:
    print('Invalid value imput')
else:
    print('Valid user input')
```

```python
#Finally block

try:
    f = open('data.txt')
except FileNotFoundError:
    print('File not found')
finally:
    print('Closing file......')
```

```python
#try-except-else-finally

try:
    value = int(input('Enter a value:'))
    result = 10/value

except ZeroDivisionError:
    print('Value can\'t be zero')
except ValueError:
    print('Enter only integer value')
else:
    print(f'Result: {result}')
finally:
    print('Program finished')
```

```python
#try, raise error, except, finally
#Need to throw an error if the withdrawal amount from the atm is more than the user's balance amount

try:
    account_balance = 12000
    withdrawal_amount = int(input('Enter the amount to withdraw:'))
    if withdrawal_amount > account_balance:
        raise Exception('Insufficient Balance ðŸ˜¢')
    elif withdrawal_amount <= 100:
        raise Exception('Withdrawal amount should be more than Rs.100')
    else:
        final_balance = account_balance - withdrawal_amount
        print(f'You have withdrawan: {withdrawal_amount}')
except ValueError:
    print('Enter only integer value')
    final_balance = account_balance
except Exception as e:
    print(e)
    final_balance = account_balance
finally:
    print(f'Your account balance: {final_balance}')
```
