```python
name = input('Enter your name:').lower()
vowels = 'aeiou'

count = 0

for char in name:
    if char in vowels:
        count +=1
print(f'Number of vowels in your name is: {count}')
```

```python
# Task

sent = '     This is innomatics and we provide Data Science course       '

#upper, strip, split, replace (science to analysis)

sent_upper = sent.strip().upper()
sent_split = sent.strip().split()
sent_strip = sent.strip()
sent_replace = sent.strip().replace('Science', 'Analysis')

#Output
print(f'Upper case: {sent_upper}')
print(f'Strip: {sent_strip}')
print(f'Split: {sent_split}')
print(f'Replace: {sent_replace}')
```
