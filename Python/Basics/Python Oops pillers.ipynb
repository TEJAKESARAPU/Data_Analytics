```python
# An abstract class can have both abstract methods and normal methods
# A child class must implement all the abstract methods
# Cannot create an object of an abstract class
```

```python
from abc import ABC, abstractmethod

class vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass    # abstract method

    def display(self):
        print('This is a vehicle')   # normal method


class car(vehicle):   # child class
    def start(self):
        print('My car engine starts with key!')


my_car = car()
my_car.start()
my_car.display()
```

```python
# Create an abstract class for an animal
# Create two child classes to get the voice of the animal

from abc import ABC, abstractmethod

class animal(ABC):

    @abstractmethod    #abstract class
    def voice(self):
        pass
        
class cat(animal):     #child class
    def voice(self):
        print('Cat purrs as meow-meow')

class dog(animal):     #child class
    def voice(self):
        print('Dog barks as bow-bow')

cat = cat()
dog = dog()

cat.voice()
dog.voice()
```

```python
# # payment
# # create child class 
# # payment by cc and UPI and online transaction
# define abstract class and child class
# design a code for an abstract class
    
    from abc import ABC, abstractmethod
    
    class payments(ABC):
    
        @abstractmethod          #abstract method
        def transaction(self):
            pass
    
    class cc(payments):         #child class
        def transaction(self):
            return f'Your payment has been processed via credit card'
    
    class upi(payments):        #child class
        def transaction(self):
            return f'Your payment has been processed via UPI'
    
    cc = cc()
    upi = upi()
    print(cc.transaction())
    print(upi.transaction())
```

```python
class student:
    def __init__(self):
        self.name = 'mohit'   #public variable
        self._age = 28        #protected variable (_)

s1 = student()
print(s1.name)
print(s1._age) #technically okay but not recommended
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance   #private variable

    def deposit(self, amount):
        self.__balance +=amount

    def get_balance(self):
        return self.__balance

acc1 = BankAccount()
acc1.deposit(5000)
print(acc1.get_balance())
```

```python
class employee:

    def __init__(self, salary=0):
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, amount):
        if amount > 0:
            self.__salary = amount
        else:
            print('Invalid salary')

#defining object

emp1 = employee(3000)
print(emp1.get_salary())
```

```python
emp1.set_salary(20000)
print(emp1.get_salary())
```
