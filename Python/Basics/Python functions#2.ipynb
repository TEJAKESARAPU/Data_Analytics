```python
''' args - Arbitrary arguments '''
```

```python
def numbers(*args):
    print(args)

numbers(10,20,30,40)
```

```python
def pri(*prices):
    return sum(prices)
pri(100,200,300,400)
```

```python
def total_price(*prices):
    total = 0
    for price in prices:
        total += price
    return total

total_price(100,200,300,400)
```

```python
''' ** kw args - Arbitrary keyword argument '''
```

```python
def student_info(**kwargs):
    print (kwargs)
```

```python
student_info(name = 'ram', age = 25, gender = 'male')
```

```python
def create_profile(**details):
    for key, value in details.items():
        print(f'{key} : {value}')
```

```python
create_profile(name = 'Teja', age = 24, city = 'Visakhapatnam')
```

```python
def my_func(**kwargs):
    name = kwargs.get('name') #['name']
    age = kwargs.get('age')
    print(f'Hello! My name is {name} and my age is {age}')
```

```python
my_func(name = 'Teja', age = 24)
```

```python
def my_function(username, **details):
    print('Username:', username)
    print('Additional details:')
    for key, value in details.items():
        print(' ', key + ':', value)
```

```python
my_function('tejagiri', name = 'Teja', age = 24, city = 'Visakhapatnam')
```

```python
def student_details(roll_no, *marks, **info):
    print('Student roll number:', roll_no)
    print('Student marks:', marks)
    print('Student info:', info)
```

```python
student_details(102983106, 
                90,85,75,97, 
                name = 'Teja', age = 24, school = 'Chalapathi High School')
```

```python
''' Lambda function '''
```

```python
square = lambda x: x*x
square(5)
```

```python
add = lambda a,b: a+b
add(5,4)
```

```python
even = lambda x: x%2==0
even(5)
```

```python
''' FIlter function '''
```

```python
ages = [12,30,18,24,35,17,55]
adults = list(filter(lambda age: age >= 18, ages))
print('Adults:', adults)
```

```python
''' Map function '''
```

```python
num = [1,2,3,4,5]
square = list(map(lambda x: x*x, num))
print(square)
```

```python
prices = [100, 150, 170, 200]
discount = list(map(lambda x: x - (x*0.10), prices))
print(discount)
```

```python
prices = [100, 150, 170, 200]
gst = list(map(lambda x: x + (x*0.18), prices))
print(gst)
```

```python
''' Reduce function '''
```

```python
from functools import reduce

prices = [100,200,300,400]
total = reduce(lambda a,b : a+b, prices)
print(total)
```
