```python
#duck typing method

class dog:
    def sound(self):
        return f'Bow-Bow'

class cat:
    def sound(self):
        return f'Meow-Meow'

def make_sound(animal):
    print(animal.sound())

make_sound(dog())
make_sound(cat())
```

```python
#method overloading
#using default args.

class mathops:
    def add(self, a,b=0, c=0):
        return a + b + c

m = mathops()
print(m.add(5))
print(m.add(5,10))
print(m.add(5,10,15))
```

```python
#using *args

class calculator:
    def add(self, *nums):
        return sum(nums)

cal = calculator()
print(cal.add(1,2,3,4))
```

```python
#method overriding

class vechicle:
    def start(self):
        print('Staring the vehicle..........')

class car(vechicle):
    def start(self):
        print('Car starts with push start button')

class bike(vechicle):
    def start(self):
        print('Bike start with a self start button')

vechicle().start()
car().start()
bike().start()
```

```python
#Operator overloading
#overloading '+' operator

class point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return point(self.x + other.x , self.y + other.y)

p1 = point(2,3)
p2 = point(4,5)
p3 = p1 + p2
print(p3.x, p3.y)
```

```python
#overloading '>' operator

class box:
    def __init__(self, volume):
        self.volume = volume

    def __gt__(self, other):
        return self.volume > other.volume

b1 = box(45)
b2 = box(60)

print(b2 > b1)
```

```python
#polymorphism with functions & classes

class sparrow:
    def fly(self):
        return f'Sparrow flying low'

class eagle:
    def fly(self):
        return f'Eagle flying high'

def show_flight(bird):
    print(bird.fly())

show_flight(sparrow())
show_flight(eagle())
```

```python
# classroom task
# class - payment
# method - pay
# child classes - 3-4

class payment:
    def pay(self):
        pass

class ccpayment(payment):
    def pay(self):
        return f'Your payment has been processed with the Credit Card'

class upipayment(payment):
    def pay(self):
        return f'Your payment has been processed with the UPI payment'

class netbanking(payment):
    def pay(self):
        return f'Your payment has been processed with the NetBanking'

class wallet(payment):
    def pay(self):
        return f'Your payment has been processed with the wallet'

def show_payment(payments):
    print(payments.pay())

show_payment(ccpayment())
show_payment(upipayment())
show_payment(netbanking())
show_payment(wallet())
```
