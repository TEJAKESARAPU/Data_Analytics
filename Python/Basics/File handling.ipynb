```python
#creating a file

with open('student.txt', 'w') as f:
    f.write('Name: Teja\n')
    f.write('Age: 24\n')
    f.write('Course: Data Analytics\n')
```

```python
#Adding more info

with open('student.txt', 'a') as f:
    f.write('City: Visakhapatnam\n')
    f.write('Grade: A\n')
```

```python
#reading a file

with open('student.txt', 'r') as f:
    print(f.read())
```

```python
#reading a file with limited characters

with open('student.txt', 'r') as f:
    print(f.read(10))
```

```python
#read line by line

with open('student.txt', 'r') as f:
    print(f.readline())
    print(f.readline())
    print(f.readline())
```

```python
#reading all the lines as a list

with open('student.txt', 'r') as f:
    print(f.readlines())
```

```python
#loop through a file

with open('student.txt', 'r') as f:
    for line in f:
        print('-->', line.strip())
```

```python
#Read and write together

with open('student.txt', 'r+') as f:
    data = f.read()
    f.write('This is the student record\n' + data)
    print(data)
```

```python
with open('student.txt', 'r') as f:
    print(f.read())
```

```python
#write and ready together

with open('python.txt', 'w+') as f:
    f.write('This is Python file\n')
    f.seek(0)   #cursor position
    print(f.read())
```

```python
#binary file

with open('photo.jpg', 'rb') as file1:
    data = file1.read()

with open('copy.jpg', 'wb') as file2:
    file2.write(data)
```
