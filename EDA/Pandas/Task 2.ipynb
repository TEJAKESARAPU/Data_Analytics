```python
# Task - Create a dataframe including emp_id, emp_name and emp_salary.
# Get single and multiple columns
# Swap the columns
# Convert all the columns into lower and upper case 
# Add new columns based on some condition
# Delete all the newly added columns
# Give the 10% increment to all the employees
```

```python
import pandas as pd
```

```python
# Creating the data frame

emp_data = {
    'emp_id': [101,102,103,104,105,106],
    'emp_name': ['Teja', 'Ram', 'Sam', 'Avi', 'Tara', 'Anu'],
    'emp_salary': [50000,45000,35000,41500,31640,46660]
}
df = pd.DataFrame(emp_data)
df
```

```python
# Get single and multiple columns

print(f'Single column:\n{df.emp_id}')
print(f'\nMulyiple columns:\n{df[['emp_name', 'emp_salary']]}')
```

```python
# Swap the columns

df = df[['emp_name', 'emp_id', 'emp_salary']]
df
```

```python
# Convert all the columns into lower and upper case 

df.columns = df.columns.str.upper()  #Upper case
df
```

```python
df.columns = df.columns.str.lower()  #Lower case
df
```

```python
# Add new columns based on some condition

df[df['emp_salary'] > 43000]
```

```python
# Delete all the newly added columns

df.drop(['emp_salary'], axis=1)
```

```python
# Give the 10% increment to all the employees

df['emp_salary'] = df['emp_salary'] * 1.10
```
