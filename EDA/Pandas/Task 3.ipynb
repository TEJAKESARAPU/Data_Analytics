```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

bank_data = {
    'BankName': ['HDFC', 'ICICI', 'AXIS', 'CUB'],
    'IFSC': np.random.randint(120001000, 120002000, 4),
    'AccountHolder': ['Ram', 'Teja', 'Sam', 'Mohan'],
    'Amount': [12000,85000,5000,46000]
}
```

```python
df = pd.DataFrame(bank_data)
```

```python
df
```

```python
# getting rows (df['rows'], df.index, loc, iloc, at, iat)

print(df['IFSC'],'\n')
print(df.index, '\n')
print(list(df.index), '\n')
print(df.loc[1:2, 'IFSC':], '\n')
print(df.iloc[2:,2:], '\n')
print(df.at[1, 'AccountHolder'],'\n')
print(df.iat[2,1])
```
