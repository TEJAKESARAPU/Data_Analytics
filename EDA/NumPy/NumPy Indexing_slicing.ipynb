```python
#Boolean slicing

arr = np.array([10,20,30,40,50])
print(arr)
```

```python
print(arr[arr>30])
```

```python
print(arr[(arr>30) & (arr<50)])
```

```python
arr[3] = 400
print(arr)
```

```python
arr[arr>30] = 0
print(arr)
```
