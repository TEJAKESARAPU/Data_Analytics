```python
import numpy as np
from scipy.stats import skew
```

```python
data1 = [16, 14, 22, -11, 16, 23, 25, 19, 12, 27, 21, 14, 16, 85, 13, 14, 16, 19, 14, 16]
data2 = [18, 21, 12, 24, -17, 17, 20, 26, 11, 14, 23, 17, 16, 12, 18, 17, 18, 21, 18, 17]
data3 = [9, 15, 22, 11, -320, 19, 24, 20, 17, 23, 16, 12, 18, 18, 18, 15, 15, 22, 15, 18]
data4 = [18, 10, 26, 14, 22, 24, -25, 21, 13, 12, 16, 19, 18, 18, 14, 14, 18, 16, 14, 18]
data5 = [20, -70, 14, 22, 11, 23, 25, 17, -40, 19, 15, 14, 14, 14, 14, 19, 19, 15, 14, 19]

datasets = {
    'data1': data1,
    'data2': data2,
    'data3': data3,
    'data4': data4,
    'data5': data5
}

for name, data in datasets.items():
    s = skew(data)
    print(f'{name} skewness: {s}')
    if s > 0:
        print('-> Positively skewed (Right skewed)\n')
    elif s<0:
        print('-> Negatively skewed (Left skewed)\n')
    else:
        print('-> Symmtric\n')
```
