##### 2. Write a NumPy program to check whether two arrays are equal (element wise) or not.

###### 3. Write a NumPy program to create a 4x4 array with random values, now create a new array from the said array swapping first and last rows.

```python
arr = np.arange(0,16).reshape(4,4)
print(f'Origina array:\n{arr}\n')
arr_swapped = arr.copy()
arr_swapped[[0,-1]] = arr_swapped[[-1,0]]
print(f'New array after swapping first and last rows of the said array:\n{arr_swapped}')
```

###### 4. Write a NumPy program to convert a list and tuple into arrays.

```python
lst = [1,2,3,4,5]
tup = (6,7,8,9,10)
arr1 = np.array(lst)
arr2 = np.array(tup)
print(f'List to array: {arr1}')
print(f'Tuple to array: {arr2}')
```

###### 5. Write a NumPy program to find common values between two arrays.

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([5,6,7,1,8])
arr_common = np.intersect1d(arr1,arr2)

print(f'First array: {arr1}\nSecond array: {arr2}')
print(f'Common values between two arrays: {arr_common}')
```

###### 7. Write a NumPy program to count the frequency of unique values in numpy array.

```python
arr = np.array([10,10,20,10,20,20,20,30,30,50,40,40])
print(f'Original array:\n{arr}')

unique_values, counts = np.unique(arr, return_counts=True)
frequency = np.array([unique_values, counts])

print(f'Frequency of unique values of the said array:\n{frequency}')
```

###### 11. Write a NumPy program to get the dates of yesterday, today and tomorrow.

```python
today = np.datetime64('today', 'D')
yesterday = today - np.timedelta64(1, 'D')
tomorrow = today + np.timedelta64(1, 'D')
print(f'Today: {today}\nYesterday: {yesterday}\nTomorrow: {tomorrow}')
```

###### 12. Write a NumPy program to find the first Monday in May 2017.

```python
may2017 = np.datetime64('2017-05')
first_monday = np.busday_offset(may2017, 0, roll='forward', weekmask='Mon')
print(f'First Monday in May 2017:\n{first_monday}')
```
