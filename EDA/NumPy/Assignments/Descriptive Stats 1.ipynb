```python
# data1 = [16, 14, 22, -11, 16, 23, 25, 19, 12, 27, 21, 14, 60, 85, 13]
# data2 = [18, 21, 12, 24, -17, 17, 20, 26, 11, 14, 23, 75, 16, 1210, 12]
# data3 = [9, 15, 22, 11, -320, 19, 24, 20, 17, 23, 16, 12, 1455, 18, 180]
# data4 = [18, 10, 26, 14, 22, 24, -25, 21, 13, 12, 90, 16, 19, 160, 12]
# data5 = [20, -70, 14, 22, 11, 23, 25, 17, -40, 19, 15, 95, 9, 1840, 14]

# Tasks : Calculate the following measures for the above data
# Mean
# Median
# Variance
# Standard Deviation
# IQR
# Find the lower fence and upper fence and identify if there are any outliers.

import numpy as np

data1 = [16, 14, 22, -11, 16, 23, 25, 19, 12, 27, 21, 14, 60, 85, 13]
data2 = [18, 21, 12, 24, -17, 17, 20, 26, 11, 14, 23, 75, 16, 1210, 12]
data3 = [9, 15, 22, 11, -320, 19, 24, 20, 17, 23, 16, 12, 1455, 18, 180]
data4 = [18, 10, 26, 14, 22, 24, -25, 21, 13, 12, 90, 16, 19, 160, 12]
data5 = [20, -70, 14, 22, 11, 23, 25, 17, -40, 19, 15, 95, 9, 1840, 14]

datasets = {
    'data1': data1,
    'data2': data2,
    'data3': data3,
    'data4': data4,
    'data5': data5
}

for name, data in datasets.items():
    print(f'\n==== {name} ====\n')

    mean_val = np.mean(data)
    median_val = np.median(data)
    var_val = np.var(data)
    std_val = np.std(data)
    
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3 - q1

    lower_fence = q1 - 1.5 * IQR
    upper_fence = q3 + 1.5 * IQR

    outliers = [x for x in data if x < lower_fence or x > upper_fence]

    print(f'Mean: {mean_val}')
    print(f'Median: {median_val}')
    print(f'Variance: {var_val}')
    print(f'Standard deviation: {std_val}')
    print(f'Q1: {q1}')
    print(f'Q3: {q3}')
    print(f'IQR: {IQR}')
    print(f'Lower Fence: {lower_fence}')
    print(f'Upper Fence: {upper_fence}')
    print(f'Outliers: {outliers}')
```
